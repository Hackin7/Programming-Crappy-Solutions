```python
class ListNode:
  def __init__(self, dataValue, pointerValue):
    self.dataValue = dataValue
    self.pointerValue = pointerValue

class LinkedList:
  def __init__(self):
    self.nodes = [0]*31
    for i in range(1,31): # Start from index 1
      pointerValue = i+1
      self.nodes[i] = ListNode("empty", pointerValue)
    self.nextFree = 1
  
  def isEmpty(self) -> bool:
    """
    If the first element is empty, there is no pointer to other elements of the list
    If first element not empty, next element is empty but the pointerValue is not
    properly initialised, it shows that the first line of addNode ran and that
    the list was empty before running addNode 
    """

    return self.nodes[1].dataValue == "empty" or \
    (self.nodes[1].pointerValue == 2 and self.nodes[2].dataValue == "empty") 

  def addNode(self, newData):
    self.nodes[self.nextFree].dataValue = newData
    if self.isEmpty() == True:
      print("Empty")
      temp = self.nodes[self.nextFree].pointerValue
      self.nodes[self.nextFree].pointerValue = 0 # End of List
      self.nextFree = temp
    else:
      temp = self.nodes[self.nextFree].pointerValue
      self.nodes[self.nextFree-1].pointerValue = self.nextFree
      self.nodes[self.nextFree].pointerValue = 0 # End of List
      self.nextFree = temp
```

```python
# Driver Code
lis = LinkedList()
def printLis():
  output=""
  for n in lis.nodes[1:21]:
    output += str(n.dataValue) + ","
  print(output)


print(lis.isEmpty())
printLis()
print(lis.nextFree)

#Test addNode
lis.addNode("Mango")
lis.addNode("Orange")
lis.addNode("Banana")
lis.addNode("Lemon")
print(lis.isEmpty())
printLis()
print(lis.nextFree)
```
