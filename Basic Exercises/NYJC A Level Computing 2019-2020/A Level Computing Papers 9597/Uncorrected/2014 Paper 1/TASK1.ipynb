```python
# Task 1.

def readFile(filename="JARGON.TXT"):
    array = [] # Python list to simulate an array
    file = open(filename, 'r')
    for line in file:
        data = line.strip()
        array.append(data)
    file.close()
    return array

## Options #########################################################
def exactMatch(item, term): # Option 1
    return item == term

def startsWith(item, term): # Option 2
    return len(item) >= len(term) and item[:len(term)] == term

def withinTerm(item, term):
    return term in item

optToCmp = {
    1: exactMatch,
    2: startsWith,
    3: withinTerm
}
## Menu ##########################################################
def matching(array,cmp,term):
    output = ""
    counter = 0
    for item in array:
        if cmp(item, term):
            output += f"{item}\n"
            counter += 1
    output += f"There were {counter} matching term(s)"
    return output

def menu():
    array = readFile()
    run = True
    while run:
        # Menu text and Input
        print("+++++++++++++++++++++++")
        print("1. Exact Match")
        print("2. Start of term")
        print("3. Within term")
        print("++++++++++++++++++")
        choice = int(input("Choice ?"))
        term = input("Term?")
        
        if term == "XXX":
            run = False
        elif choice in optToCmp.keys():
            print(matching(array, optToCmp[choice], term))
            print()
            
menu()
```
