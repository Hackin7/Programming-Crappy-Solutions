```python
# Task 3.1
MENU_TEXT = """\
1. Add an item
2. Traverse the linked list of used nodes and output the data values
3. Output all pointers and data values
4. Traverse the linked list IN REVERSE of used nodes and output the data values
5. Exit\
"""
def menu(linkedlist):
    run = True
    while run:
        print(MENU_TEXT)
        option = input("Enter choice: ").strip()
        if option == "1":
            linkedlist.AddNode()
        elif option == "2":
            linkedlist.Traversal()
        elif option == "3":
            linkedlist.DisplayLinkedList()
        elif option == "4":
            linkedlist.ReverseTraversal()
        elif option == "5":
            run = False
```

```python
# Task 3.2
class ListNode:
    def __init__(self, DataValue: str, PointerValue: int):
        self.DataValue = str(DataValue)
        self.PointerValue = int(PointerValue)
        
class LinkedList:
    def __init__(self):
        self.Initialise()
        
    def Initialise(self):
        # Use a python list to simulate an array
        self.Node = []
        size = 30
        for i in range(size + 1): # 1-indexed
            nextPos = (i + 1) % size
            self.Node.append(ListNode('', nextPos))
        # Others
        self.Start = 0
        self.NextFree = 1
        
    def DisplayLinkedList(self):
        '''
        Display current state of pointers and array contents
        '''
        output = ""
        # Pointers
        output += f"Start = {self.Start}"
        output += f"\nNextFree = {self.NextFree}"
        
        # Array contents
        output += f"\n{'Index':5} {'DataValue':20} {'PointerValue':15}"
        for index in range(1, len(self.Node)):
            output += f"\n{index:5} {self.Node[index].DataValue:20} {self.Node[index].PointerValue:15}"
            
        print(output)
    
    def isEmpty(self) -> bool:
        return self.Start == 0
```

```python
# Task 3.3
def driver():
    linkedlist = LinkedList()
    menu(linkedlist)

driver()
```

```python
# Task 3.4
class LinkedList(LinkedList):
    def AddNode(self):
        NewItem = input("Enter Item to add: ")
        self.Node[self.NextFree].DataValue = NewItem

        if self.Start == 0: 
            self.Start = self.NextFree
            # Swapping
            Temp = self.Node[self.NextFree].PointerValue
            self.Node[self.NextFree].PointerValue = 0
            self.NextFree = Temp
        else:
            # traverse the list â€“ starting at Start to find
            # the position at which to insert the new item
            Temp = self.Node[self.NextFree].PointerValue

            if NewItem < self.Node[self.Start].DataValue:
                # new item will become the start of the list
                self.Node[self.NextFree].PointerValue = self.Start
                self.Start = self.NextFree
                self.NextFree = Temp
            else:
                # the new item is not at the self.Start of the list . . .
                Previous = 0
                Current = self.Start
                Found = False
                
                untilCondition = False
                while not untilCondition:
                    if NewItem <= self.Node[Current].DataValue:
                        self.Node[Previous].PointerValue = self.NextFree
                        self.Node[self.NextFree].PointerValue = Current
                        self.NextFree = Temp
                        Found = True
                    else:
                        # move on to the next self.Node
                        Previous = Current
                        Current = self.Node[Current].PointerValue
                    
                    ## Until condition
                    untilCondition = (Found == True or Current == 0)

                if Current == 0:
                    self.Node[Previous].PointerValue = self.NextFree
                    self.Node[self.NextFree].PointerValue = 0
                    self.NextFree = Temp

'''
Test Data:
1
MANGO
1
ORANGE
1
BANANA
1
LEMON
3
5
'''
driver()
```

```python
# Task 3.5

# Procedure
def TraversalInOrder(Index, Node=None):
    if Index != 0:
        print(Node[Index].DataValue)
        # Follow pointer to next data item in the linked list
        TraversalInOrder(Node[Index].PointerValue, Node)

class LinkedList(LinkedList):
    def Traversal(self):
        TraversalInOrder(self.Start, self.Node)
        
driver()
```

```python
# Task 3.6

# Procedure
def ReverseTraversalInOrder(Index, Node=None):
    if Index != 0:
        # Follow pointer to next data item in the linked list
        TraversalInOrder(Node[Index].PointerValue, Node)
        # Print AFTER the recursive call
        print(Node[Index].DataValue)

class LinkedList(LinkedList):
    def ReverseTraversal(self):
        ReverseTraversalInOrder(self.Start, self.Node)
        
driver()
```
