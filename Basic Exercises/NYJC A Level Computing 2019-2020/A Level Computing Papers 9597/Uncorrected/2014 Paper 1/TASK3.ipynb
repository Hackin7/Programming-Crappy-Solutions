```python
# Task 3.1
MENU_TEXT = """\
1. Add an item
2. Traverse the linked list of used nodes and output the data values
3. Output all pointers and data values
4. Traverse the linked list IN REVERSE of used nodes and output the data values
5. Exit\
"""
def menu(linkedlist):
    run = True
    while run:
        print(MENU_TEXT)
        option = input("Enter choice: ").strip()
        if option == "1":
            linkedlist.AddNode()
        elif option == "2":
            linkedlist.Traversal()
        elif option == "3":
            linkedlist.DisplayLinkedList()
        elif option == "4":
            linkedlist.ReverseTraversal()
        elif option == "5":
            run = False
```

```python
# Task 3.2
class ListNode:
    def __init__(self, DataValue: str, PointerValue: int):
        self.__DataValue = str(DataValue)
        self.__PointerValue = int(PointerValue)
        
    def getDataValue(self):
        return self.__DataValue
    def getPointerValue(self):
        return self.__PointerValue
    
    def setDataValue(self, DataValue):
        self.__DataValue = str(DataValue)
    def setPointerValue(self, PointerValue):
        self.__PointerValue = int(PointerValue)
        
        
class LinkedList:
    def __init__(self):
        self.Initialise()
        
    def Initialise(self):
        # Use a python list to simulate an array
        self.Node = []
        size = 30
        for i in range(size + 1): # 1-indexed
            nextPos = (i + 1) % (size + 1)
            self.Node.append(ListNode('', nextPos))
        # Others
        self.Start = 0
        self.NextFree = 1
        
    def DisplayLinkedList(self):
        '''
        Display current state of pointers and array contents
        '''
        output = ""
        # Pointers
        output += f"Start = {self.Start}"
        output += f"\nNextFree = {self.NextFree}"
        
        # Array contents
        output += f"\n{'Index':5} {'DataValue':20} {'PointerValue':15}"
        for index in range(1, len(self.Node)):
            output += f"\n{index:5} {self.Node[index].getDataValue():20} {self.Node[index].getPointerValue():15}"
            
        print(output)
    
    def isEmpty(self) -> bool:
        return self.Start == 0
```
