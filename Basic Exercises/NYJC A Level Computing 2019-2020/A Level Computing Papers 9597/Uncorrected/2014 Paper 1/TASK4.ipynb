# Task 4.1
```
Data structure to represent the grid: 2D Array of characters (string) (8,15)
Contents of each square meter per pond:
    '.' for 1 square meter of water
    'S' for Stone impact position
    
Program Design:
                 [Pond]
                   |
      ---------------------------
      |            |            |             
[createPond]  [addStone] [Display Pond]
                   |
            [enterCoordinates]
```

````python
# Task 4.2

def createPond(X=15, Y=8):
    # Use a python list to simulate an array
    array = [] # 1-indexed, 0 indexes to be ignoresd
    for y in range(Y + 1):
        row = []
        for x in range(X + 1):
            row.append('.')
        array.append(row)
    return array


def enterCoordinates(array):
    X = len(array[0]) - 1
    Y = len(array) - 1
    x = int(input(f"X coordinate <1 to {X}>? "))
    y = int(input(f"Y coordinate <1 to {Y}>? "))
    return x, y
    
def addStone(array, x, y):
    array[y][x] = 'S'

def displayPond(array):
    output = ""
    # Displaying main element
    for row in array[1:]: # Slicing to exclude first element
        for pos in row[1:]:
            output += f"{pos}"
        output = output + "\n"
    # Remove trailing new line
    output = output[:-1] 
    print(output)
        
def task4_2():
    array = createPond()
    x, y = enterCoordinates(array)
    addStone(array, x, y)
    displayPond(array)
    
task4_2()
````

````python
import random
def addFish(array, no=3):
    X = len(array[0]) - 1
    Y = len(array) - 1
    # Generate list of possible positions
    positions = []
    for x in range(1, X + 1): # 1-indexed
        for y in range(1, Y + 1):
            positions.append((x,y))
    
    # Randomly choose positions
    while no > 0 and len(positions) >= 0:
        posIndex = random.randint(0, len(positions) - 1)
        x, y = positions[posIndex]
        array[y][x] = 'F'
        positions.pop(posIndex)
        no -= 1

def task4_3():
    array = createPond()
    addFish(array)
    displayPond(array)
    
task4_3()
````

````python
# Task 4.4
def displayPond(array, showF=False):
    output = ""
    # Displaying main element
    for row in array[1:]: # Slicing to exclude first element
        for pos in row[1:]:
            if pos == 'F' and not showF:
                item = '.'
            else:
                item = pos
            output += f"{item}"
        output = output + "\n"
    # Remove trailing new line
    output = output[:-1] 
    print(output)
    
def addPellet(array, x, y):
    if array[y][x] == 'F':
        array[y][x] = 'H' # Happy fish
    else:
        array[y][x] = 'P' # Pellet
        
def getCoordinates(array):
    coordinates = []
    for row in range(1, len(array)):
        for col in range(1, len(array[row])):
            if array[row][col] == 'F':
                coordinates.append((col, row))
    print(coordinates)
    
def task4_4():
    array = createPond()
    addFish(array)
    getCoordinates(array)
    print()
    print("Before")
    displayPond(array, False)
    print("Add pellet")
    x, y = enterCoordinates(array)
    addPellet(array, x, y)
    print()
    displayPond(array, True)
    
for i in range(2):
    print("### Test Case " + str(i+1) + " " + "#" * 30)
    task4_4()
    print()
    
````
