```python
# Task 3.1
class Node:
    def __init__(self, LeftP: int, Data: str, RightP: int):
        self.LeftP = int(LeftP)
        self.Data = str(Data)
        self.RightP = int(RightP)
        
class BinaryTreeList: # Binary Tree with Unused Nodes in Linked List
    def __init__(self):
        # Python List to simulate an array
        size = 20
        self.ThisTree = [] 
        for i in range(size + 1): # 1-indexed: 0-index not put to use
            nextPos = (i + 1) % (size + 1)
            self.ThisTree.append(Node(nextPos , '', 0))
        
        # Other Variables
        self.Root = int(0)
        self.NextFreePosition = int(1)
        
bintreelist = BinaryTreeList()
```

```python
# Task 3.2

class BinaryTreeList(BinaryTreeList):
    def AddItemToBinaryTree(self, NewTreeItem):
        ### Test for space of new item #########################################
        if self.NextFreePosition == 0: # If it ran out of space
            return 
        ### Add Item to Tree List ##############################################
        self.ThisTree[self.NextFreePosition].Data = NewTreeItem
        nextFree = self.ThisTree[self.NextFreePosition].LeftP
        self.ThisTree[self.NextFreePosition].LeftP = 0 # Unlink
        ########################################################################
        
        if self.Root == 0 : # 0
            self.Root = self.NextFreePosition
        else:
            ### Traverse the tree to find the position for the new value ##################
            CurrentPosition = self.Root
            LastMove = 'X'

            ## Simulate a REPEAT UNTIL Loop
            untilCondition = False
            while not untilCondition:
                PreviousPosition = CurrentPosition
                if NewTreeItem < self.ThisTree[CurrentPosition].Data:
                    # Move left
                    LastMove = 'L'
                    CurrentPosition = self.ThisTree[CurrentPosition].LeftP
                else:
                    # Move right
                    LastMove = 'R'
                    CurrentPosition = self.ThisTree[CurrentPosition].RightP
                untilCondition = (CurrentPosition == 0)
            ##############################################################################
            
            
            if LastMove == 'R':
                self.ThisTree[PreviousPosition].RightP = self.NextFreePosition
            else:
                self.ThisTree[PreviousPosition].LeftP = self.NextFreePosition
            
        
        self.NextFreePosition = nextFree # self.ThisTree[self.NextFreePosition].LeftP 
```

```python
# Task 3.3

class BinaryTreeList(BinaryTreeList):
    def OutputData(self):
        output = ""
        
        # Pointers
        output += f"Root = {self.Root}"
        output += f"\n"
        output += f"NextFreePosition = {self.NextFreePosition}"
        
        # Array
        output += f"\n"
        output += f"ThisTree: "
        output += f"\n"
        output += f"{'Index':5} {'LeftP':5} {'Data':10} {'RightP':6}"
        for index in range(1, len(self.ThisTree)):
            node = self.ThisTree[index]
            output += f"\n"
            output += f"{index:5} {node.LeftP:5} {node.Data:10} {node.RightP:6}"
        
        print(output)
        
```
