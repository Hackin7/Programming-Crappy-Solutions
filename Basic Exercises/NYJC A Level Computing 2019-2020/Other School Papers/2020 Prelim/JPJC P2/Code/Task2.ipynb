```python
#Task 2.1

class Queue:
    def __init__(self):
        self.Items = [""] * 5
        self.Front = -1
        self.Rear = -1
    
    def IsEmpty(self):
        return self.Front == -1
    
    def IsFull(self):
        return self.Rear == 4
    
    def Enqueue(self, string):
        if self.IsFull():
            print("Queue is full!")
        else:
            if self.Front == -1:
                self.Front += 1
            self.Rear += 1
            self.Items[self.Rear] = string
            
    def Dequeue(self):
        if self.IsEmpty():
            return "NONE"
        else:
            item = self.Items[self.Front]
            if self.Front == self.Rear: #last item removed
                self.Front, self.Rear = -1, -1
            else:
                self.Front += 1
            return item
        
    def Display(self):
        if self.IsEmpty():
            print("Queue is empty!")
        else:
            for i in range(self.Front, self.Rear+1):
                print(self.Items[i])
        print()

class CircularQueue(Queue):
    def __init__(self):
        super().__init__()
        
    def IsFull(self):
        return (self.Rear+1)%5 == self.Front
    
    def Enqueue(self, string):
        if self.IsFull():
            print("Queue is full!")
        else:
            if self.Front == -1:
                self.Front += 1
            self.Rear = (self.Rear+1)%5
            self.Items[self.Rear] = string
            
    def Dequeue(self):
        if self.IsEmpty():
            return "NONE"
        else:
            item = self.Items[self.Front]
            if self.Front == self.Rear: #last item removed
                self.Front, self.Rear = -1, -1
            else:
                self.Front = (self.Front+1)%5
            return item
        
    def Display(self):
        if self.IsEmpty():
            print("Queue is empty!")
        else:
            index = self.Front
            if index != self.Rear:
                while True:
                    print(self.Items[index])
                    index = (index+1)%5
                    if index == self.Rear:
                        break
            print(self.Items[index])                    
        print()        
    
```

```python
#Task 2.2

q = Queue()
cq = CircularQueue()

users = ["John","Amy","Chetan","Xin Xin","Evan"]
for user in users:
    q.Enqueue(user)
    cq.Enqueue(user)
    
print(q.Dequeue())
print(q.Dequeue())
print(cq.Dequeue())
print(cq.Dequeue())
print()
q.Enqueue("Mohan")
cq.Enqueue("Mohan")
print()
print("Linear Queue")
print("============")
q.Display()
print("Circular Queue")
print("==============")
cq.Display()
```
