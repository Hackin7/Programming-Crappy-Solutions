````python
# Task 4.1
def Cipher(NewAlphabet:list, Key:str): # list simulate Array
    Alphabet = [""] * len(NewAlphabet)
    Key = Key.upper()
    
    # Insert key first
    alphabetPos = 0
    keyIndex = 0
    while alphabetPos < len(Key) and keyIndex < len(Key):
        letter = Key[keyIndex]
        ### Check if letter inside NewAlphabet
        isInside = False
        for i in range(0, alphabetPos):
            if Alphabet[i] == letter:
                isInside = True
        ### Insert Alphabet
        if not isInside:
            Alphabet[alphabetPos] = Key[keyIndex]
            alphabetPos += 1
        keyIndex += 1
    
    index = 0
    while index < len(NewAlphabet) and alphabetPos < len(NewAlphabet):
        letter = NewAlphabet[index]
        
        ### Check if letter inside NewAlphabet
        isInside = False
        for i in range(0, alphabetPos):
            if Alphabet[i] == letter:
                isInside = True
        ### Insert remaining Alphabets
        if not isInside:
            Alphabet[alphabetPos] = letter
            alphabetPos += 1
            
        index += 1

    return Alphabet
````

````python
# Task 4.5
def readFileChars(filename="INTERCEPT.txt"):
    infile = open(filename, 'r')
    data = infile.read()
    
    # Count characters
    count = {}
    for char in data:
        if char.isalpha():
            useChar = char.upper()
            if count.get(useChar) == None:
                count[useChar] = 0
            count[useChar] += 1
    
    infile.close()
    
    # Convert dictionary to sortable list
    data = []
    for key, item in count.items():
        data.append((item, key))
    return data

def mergeSort(arr): # just for fun lmao
    ### Base Case: Already sorted ##############
    if len(arr) <= 1:
        return arr
    ### Recursive Case: sort ######################
    mid = len(arr) // 2
    left = mergeSort(arr[:mid]) # Recursive call
    right = mergeSort(arr[mid:]) # Recursive call
    ### Merging
    new = []
    while len(left) > 0 and len(right) > 0:
        if left[0][0] > right[0][0]: # left before right
            new.append(left.pop(0))
        else:
            new.append(right.pop(0))
    new = new + left + right # Add any remaining elements
    return new
    
    
def frequencyTable(data):
    # Header
    output = 'Ciphertext Letter: '
    for count, alphabet in data:
        output += f"{alphabet:2} "
    output += f'\n{"Frequency:":18} '
    for count, alphabet in data:
        output += f"{count:<2} "
    print(output)
    
def task4_5():
    data = readFileChars()
    newData = mergeSort(data)
    frequencyTable(newData)
task4_5()
````
