```python
# Task 4.1

class TreeNode:
    def __init__(self, value=None, leftPointer=None, rightPointer=None):
        self.leftPtr = leftPointer
        self.rightPtr = rightPointer
        self.value = value
        
    def getValue(self):
        return self.value
```

```python
# Task 4.2
class BSTTree:
    def __init__(self):
        self.root = None
    
    def Insert(self, value):
        if self.root == None:
            self.root = TreeNode(value=value)
        else:
            current = self.root
            while current != -1: # Terminating Case
                prev = current
                if current.getValue() < value:
                    if current.leftPtr == None: # Place to insert
                        current.leftPtr = TreeNode(value=value)
                        current = -1
                    else: # Traverse
                        current = current.leftPtr
                else:
                    if current.rightPtr == None: # Place to insert
                        current.rightPtr = TreeNode(value=value)
                        current = -1
                    else: # Traverse
                        current = current.rightPtr
        
    def ReverseOrder(self, current=None):
        if current == None: # Starting condition
            current = self.root
            print("Values in Reverse Order:")
        
        ### Displaying Values ###################
        if current.leftPtr != None:
            self.ReverseOrder(current.leftPtr)
        
        print(current.value)
        
        if current.rightPtr != None:
            self.ReverseOrder(current.rightPtr)
    
    def Search(self, value):
        if self.root == None:
            self.root = TreeNode(value=value)
        else:
            current = self.root
            # Binary Search
            while current != -1: # Terminating Case
                prev = current
                if current.getValue() < value:
                    if current.leftPtr == None:
                        return False # Does not exist
                    else: # Traverse
                        current = current.leftPtr
                elif current.getValue() > value:
                    if current.rightPtr == None: 
                        return False # Does not exist
                    else: # Traverse
                        current = current.rightPtr
                else: # elif current.getValue() == value
                    return True 
    def Count(self, current=-1):# Starting condition
        if current == -1:
            current = self.root
            
        ### Base Case: No Nodes #################
        if current == None:
            return 0
        ### Recursive Case ######################
        else:
            count = self.Count(current.leftPtr) + 1 + self.Count(current.rightPtr)
            return count
```

```python
# Task 4.3
def insert(tree, filename="INSERTTOTREE.txt"):
    file = open(filename, 'r') # Read Mode
    for itemData in file:
        item = itemData.strip()
        tree.Insert(item)
    file.close()

def output(tree):
    tree.ReverseOrder()
    print("Total Number of Nodes:", tree.Count())
    print() # New Line

def search(tree, filename="SEARCHINTREE.txt"):
    output = f"{'Item':8} {'In Tree?'}"
    file = open(filename, 'r') # Read Mode
    for itemData in file:
        item = itemData.strip()
        output += f"\n{item:8} {tree.Search(item)}"
    file.close()
    print(output)
    
### Overall Structure   
def task4_3():
    tree = BSTTree()
    insert(tree)
    output(tree)
    search(tree)

task4_3()
```
