```python
# Task 3.1
class Food:
    def __init__(self, n:str, c:str, d:str):
        self.set_name(n)
        self.set_category(c)
        self.set_description(d)
        
    ### Setter Methods ##################################
    def set_name(self, n:str):
        self.__name = str(n)
    
    def set_category(self, c:str):
        self.__category = str(c)
    
    def set_description(self, d:str):
        self.__description = str(d)
    
    ### Getter Methods #################################
    def get_name(self) -> str:
        return self.__name
    
    def get_category(self) -> str:
        return self.__category
    
    def get_description(self) -> str:
        return self.__description
    
    ### Other ##########################################
    def summary(self) -> str:
        return f"Name: {self.get_name():20} Category: {self.get_category():20} Description: {self.get_description():30}"
    
    
```

```python
# Task 3.2
class Food(Food):
    def compare_with(self, f:Food) -> int:
        currCat = self.get_category()
        fCat = f.get_category()
        if fCat < currCat:
            return 1
        elif fCat == currCat:
            currName = self.get_name()
            fName = f.get_name() 
            # Use python's string comparison for name
            if fName < currName:
                return 1
            elif fName == currName:
                return 0
            elif fName > currName: # current comes before given
                return -1
        else:
            return -1

def cmp(a, b):
    return a.compare_with(b) == 1

def insertFood(foodList, f, cmp):
    if len(foodList) == 0:
        foodList.append(f)
    else: # Traverse to appropriate spot to insert
        foodList.append(f) # Empty spot for shifting
        # Shift up while element to insert comes before element
        curr = len(foodList) - 1 - 1
        while curr >= 0 and cmp(foodList[curr], f):
            foodList[curr + 1] = foodList[curr]
            curr -= 1
        
        # Insertion
        foodList[curr + 1] = f
    
### Main ############################################
def task3_2():
    foodList = []
    toAdd = [
        Food("Apple", "Fruits", "6 Sweet Pink, bought on 23 May"),
        Food("Milk", "Diary", "Low fat, 1 litre"),
        Food("Carrot", "Vegetables", "medium sized, Australian"),
        Food("Chicken meat", "Protein", "thigh meat, 200g"),
        Food("Cabbage", "Vegetables", "300g")
    ]
    for f in toAdd:
        insertFood(foodList, f, cmp)
    
    for f in foodList:
        print(f.summary())

task3_2()
```

```python
# Task 3.3

class DatedFood(Food):
    def __init__(self, n:str, c:str, d:str, dt: str): # Use string to simulate date object
        super().__init__(n, c, d)
        self.set_expiry_date(dt)
        
    def set_expiry_date(self, d:str):
        self.__expiry_date = str(d) # YYYY-MM-DD
    
    def get_expiry_date(self):
        return self.__expiry_date
    
    def compare_with(self, f):
        # Use python string comparion to compare due date (Possible because date string is ordered from year to day)
        if type(f) == DatedFood:
            # Compare Dates
            currDate = self.get_expiry_date()
            fDate = f.get_expiry_date()
            if fDate < currDate: # Given date comes before
                return 1
            elif fDate == currDate:
                return super().compare_with(f)
            elif fDate > currDate:
                return -1
        else:
            return -1#super().compare_with(f)
        
    def summary(self):
        return f"Name: {self.get_name():20} Category: {self.get_category():20} Description: {self.get_description():30} Date: {self.get_expiry_date()}"

def cmp(a, b):
    if type(a) == Food and type(b) == DatedFood:
        return b.compare_with(a) == -1
    else:
        return a.compare_with(b) == 1
    
### Main ####################################################################
def task3_3():
    foodList = []
    toAdd = [
        Food("Apple", "Fruits", "6 Sweet Pink, bought on 23 May"),
        DatedFood("Milk", "Diary", "Low fat, 1 litre", "2020-06-17"),
        Food("Carrot", "Vegetables", "medium sized, Australian"),
        DatedFood("Chicken meat", "Protein", "thigh meat, 200g", "2020-06-15"),
        DatedFood("Egg", "Protein", "1 dozen", "2020-07-10"),
        DatedFood("Yogurt", "Diary", "Greek, 1 tub", "2020-06-17"),
        Food("Cabbage", "Vegetables", "300g")
    ]
    for f in toAdd:
        insertFood(foodList, f, cmp)
    
    for f in foodList:
        print(f.summary())

task3_3()
```
