```python
## Task 1.2

import csv
def readtoDB(coll, filename="statistic_by_country.txt"):
    file = open(filename, 'r') # Read mode
    header = None
    for line in file:
        if header == None:
            header = line
        else:
            record = line.strip().split(",")
            data = {
                "ID": record[0], 
                "Country": record[1], 
                "Region":record[2],
                "EX": record[3],
                "EW":record[4],
                "CR":record[5],
                "EN":record[6],
                "VU":record[7],
                "NT":record[8],
                "LRCD":record[9],
                "DD":record[10],
                "LC":record[11]
            }
            coll.insert_one(data)
    file.close()

readtoDB(coll)
```

```python
## Task 1.3
# Initialisation code
import pymongo
conn = pymongo.MongoClient('127.0.0.1', 27017)
db = conn.get_database("IUCN")
coll = db.get_collection("red")

header = ["ID","Country","Region","EX","EW","CR","EN","VU","NT","LRCD","DD","LC"]
spacing = [10,30,30,10,10,10,10,10,10,10,10,10,10]
keys = ["ID","Country","Region","EX","EW","CR","EN","VU","NT","LRCD","DD","LC"]

## Displaying query
def show(coll, query, header, spacing, keys):
    cursor = coll.find(query)
    #print(cursor)
    output = ""
    for index in range(len(header)):
            output += f"{header[index]:<{spacing[index]}}"
    for record in cursor:
        output += "\n"
        for index in range(len(keys)):
            key = keys[index]
            output += f"{record[key]:<{spacing[index]}}"
    print(output)

# The actual query
query = {
    "$or":[
        {"Region":"Sub-Saharan Africa"}, 
        {"Region":"East Asia"}
    ]
}
show(coll, query, header, spacing, keys)
#show(coll, {}, header, spacing, keys)
```

```python
## Task 1.4

query = {
    "$and":[
        {"NT":{
            "gt": 200
        }}, 
        {"LC":{
            "gt":1000
        }}
    ]
}
show(coll, query, header, spacing, keys)
```

```python
## Task 1.5
query = {
    "Region":{"$not":{"$eq":"South & Southeast Asia"}}, 
}
show(coll, query, header, spacing, keys)
conn.close()
```
