```python
## Task 5.1
class Staff:
    def __init__(self, name: str, appt: str, base_salary: float):
        # Name mangling to simulate private
        self.__name = str(name)
        self.__appt = str(appt)
        self.__base_salary = float(base_salary) # Monthly
    
    ### Setter #####################################
    def set_appt(self, new_appt: str):
        self.__appt = str(new_appt)
    
    def set_base_salary(self, new_base_salary: float):
        self.__base_salary = float(new_base_salary)
    
    ### Getter ####################################
    def get_name(self) -> str:
        return str(self.__name)
    
    def get_appt(self) -> str:
        return str(self.__appt)
    
    def get_base_salary(self) -> float:
        return float(self.__base_salary)
    
    def __str__(self) -> str:
        return f"{self.get_name():<20}|{self.get_appt():<20}|{self.calc_annual_salary():>12.2f}"    
    
    ### Others ####################################
    def calc_annual_salary(self) -> float:
        salary = self.get_base_salary() * 12 
        return float(salary)

class PermStaff(Staff):
    def __init__(self, name: str, appt: str, base_salary: float):
        super().__init__(name, appt, base_salary)
        self.__bonus_rating = float(0)
    
    ### Setter #####################################
    def set_bonus_rating(self, new_bonus_rating: float):
        self.__bonus_rating = float(new_bonus_rating)
    
    ### Getter ####################################
    def get_bonus_rating(self) -> float:
        return float(self.__bonus_rating)
    
    ### Others ####################################
    def calc_annual_salary(self) -> float:
        salary = self.get_base_salary() * (12 + self.get_bonus_rating()) 
        return float(salary)

class CommissionStaff(Staff):
    def __init__(self, name: str, appt: str, base_salary: float, commission_rate: float):
        super().__init__(name, appt, base_salary)
        self.__commission_rate = float(commission_rate)
        self.__revenue_generated = float(0)
    
    ### Setter #####################################
    def set_commission_rate(self, new_commission_rate: float):
        self.__commission_rate = float(new_commission_rate)
        
    def set_revenue_generated(self, new_revenue_generated: float):
        self.__revenue_generated = float(new_revenue_generated)
    
    ### Getter ####################################
    def get_commission_rate(self) -> float:
        return float(self.__commission_rate)
    
    def get_revenue_generated(self) -> float:
        return float(self.__revenue_generated)
    
    ### Others ####################################
    def calc_total_commission(self) -> float:
        return self.get_revenue_generated() + self.get_commission_rate()
    def calc_annual_salary(self) -> float:
        salary = self.get_base_salary() * 12 + self.calc_total_commission()
        return float(salary)
    
```
