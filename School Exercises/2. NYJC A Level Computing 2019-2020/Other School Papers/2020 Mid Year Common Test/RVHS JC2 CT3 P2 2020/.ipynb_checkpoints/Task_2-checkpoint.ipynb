```python
## Task 2.1
class Node:
    def __init__(self, value=None, left=None, right=None):
        self.value = value
        self.left = left
        self.right = right

class BST:        
    def __init__(self):
        self.root = None
    def insert(self, data:int):
        value = int(data)
        # Starting case
        if self.root == None:
            self.root = Node(value)
        else:
            # Traversal
            curr = self.root
            while curr != None:
                prev = curr
                if value < curr.value:
                    if curr.left == None:
                        curr.left = Node(value)
                        curr = None # Terminating case
                    else:
                        curr = curr.left
                else:
                    if curr.right == None:
                        curr.right = Node(value)
                        curr = None # Terminating case
                    else:
                        curr = curr.right
            
```

```python
## Task 2.3

# From Task 2.1
class BST:        
    def __init__(self):
        self.root = None
    def insert(self, data:int):
        value = int(data)
        # Starting case
        if self.root == None:
            self.root = Node(value)
        else:
            # Traversal
            curr = self.root
            while curr != None:
                prev = curr
                if value < curr.value:
                    if curr.left == None:
                        curr.left = Node(value)
                        curr = None # Terminating case
                    else:
                        curr = curr.left
                else:
                    if curr.right == None:
                        curr.right = Node(value)
                        curr = None # Terminating case
                    else:
                        curr = curr.right
        
    # From Task 2.2
    def inorder(self, curr=-1):
        # Initialising value
        if curr == -1:
            curr = self.root

        # Base Case
        if curr == None: 
            return
        # Recursive case
        else: 
            self.inorder(curr.left)
            print(curr.value)
            self.inorder(curr.right)

    def sum_of_longest_path(self, curr=-1):
        # Initialising value
        start = False
        if curr == -1:
            start = True
            curr = self.root

        curr_nodes = 0
        curr_sum = 0
        if curr != None:# Recursive Case
            curr_nodes = 1
            curr_sum = curr.value
            # Recursion Calls
            set1_nodes, set1_sum  = self.sum_of_longest_path(curr.left)
            set2_nodes, set2_sum = self.sum_of_longest_path(curr.right)
            
            # Comparisons
            if set1_nodes > set2_nodes or \
               (set1_nodes == set2_nodes and set1_sum > set2_sum):
                curr_nodes += set1_nodes
                curr_sum += set1_sum
            else:
                curr_nodes += set2_nodes
                curr_sum += set2_sum


        if start == True:
            return curr_sum
        else:
            # Return both values for comparison
            return (curr_nodes, curr_sum)
            
def driver():
    bst = BST()
    bst.insert(31)
    bst.insert(12)
    bst.insert(44)
    bst.insert(4)
    bst.insert(14)
    bst.insert(40)
    bst.insert(42)
    bst.insert(50)
    bst.inorder()
    print()
    print(bst.sum_of_longest_path())
driver()
```
