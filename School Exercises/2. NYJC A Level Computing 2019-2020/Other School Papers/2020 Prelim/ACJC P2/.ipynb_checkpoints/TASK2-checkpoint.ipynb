```python
# Task 2.1
import csv

### Check if move is valid ####################################
def possibleMoves(n, coordinate):
    x, y = coordinate
    moveSet = [ # x, y
        (1, 2), (1, -2), (-1, 2), (-1, -2),
        (2, 1), (2, -1), (-2, 1), (-2, -1),
    ]
    
    newPositions = []
    for dx, dy in moveSet:
        # New values
        nx = x + dx
        ny = y + dy
        if 1 <= nx <= n and 1 <= ny <= n:
            newPositions.append((nx, ny))
        
    return newPositions

def isValidMove(n, c1, c2): # coordinates
    return c2 in possibleMoves(n, c1)

### Check Visited ############################################
class VisitedMatrix:
    def __init__(self, n):
        self.matrix = []
        for y in range(n):
            row = []
            for x in range(n):
                row.append(0) # Not visited
            self.matrix.append(row)
    
    def visit(self, coordinate):
        x, y = coordinate
        self.matrix[y - 1][x - 1] += 1 # 1 indexed
        
    def allVisitedOnlyOnce(self):
        for row in self.matrix:
            for position in row:
                # If not visited
                if position != 1:
                    return False
        return True

### Check Visited ############################################
def isKnightTourFromFile(n, filename):
    visitedMatrix = VisitedMatrix(n)
    
    with open(filename, 'r') as infile:
        reader = csv.reader(infile)
        
        # Get starting Coordinate
        record = next(reader) 
        prevC = (int(record[0]), int(record[1]))
        visitedMatrix.visit(prevC)
        
        # Check if positions are valid
        for record in reader:
            currC = (int(record[0]), int(record[1]))
            if isValidMove(n, prevC, currC):
                visitedMatrix.visit(currC)
            else: 
                print("Invalid Moves:", prevC, "to", currC)
            prevC = currC
        
        infile.close() # Just in case
        
    return visitedMatrix.allVisitedOnlyOnce()

#print(possibleMoves(7, (6, 7)))
print(isKnightTourFromFile(7, "TASK2TOUR.txt"))
print(isKnightTourFromFile(7, "TASK2NOTOUR.txt"))
```
