```python
# Task 1.1

## Define the class Food
class Food:
    def __init__(self, name: str, value: int):
        self.__name = str(name)
        self.__value = int(value)
        
    def get_name(self):
        return self.__name
    
    def get_value(self):
        return self.__value
```

```python
# Task 1.2

import random
## Define the class Person
class Person:
    def __init__(self, name: str):
        self.name = str(name)
        self.health = int(100)
        self.strength = int(100)
    
    ### Getter ###########################
    def get_name(self):
        return self.name
    
    def get_health(self):
        return self.health
    
    def get_strength(self):
        return self.strength
    
    ### Others ##########################
    def eat(self, food: Food):
        self.health += food.get_value()
        # Display new strength
        print(f"New Strength of Player {self.get_name()}: {self.get_strength()}")
        
    def attack(self, opponent):
        r = random.randint(1, 10)
        if self.get_strength() < r: # Not enough strength to attack
            ### Display ###
            print(f"Attack by {self.get_name()} on {opponent.get_name()} is unsuccessful")
            print(f"    Current Strength {self.get_strength()} less than required strength {r}")
            
        elif self.get_strength() >= r: # Enough strength to attack
            ### Set Variables ###
            opponent.health -= r
            self.strength -= r # Decrease strength by r
            
            ### Display ###
            print(f"Attack by {self.get_name()} on {opponent.get_name()} is successful")
            print(f"    Current Strength dropped by {r} to {self.get_strength()} ")
            print(f"    Opponent {opponent.get_name()} health fell by {r} to {opponent.health}")
            ### Check defeated ###
            if opponent.health <= 0:
                print(f"Opponent {opponent.get_name()} is defeated")
```

```python
# Task 1.3

## Define the class Healer
class Healer(Person):
    def heal(self, patient):
        r = random.randint(1, 10)
        if self.get_strength() < r: # Not enough strength to attack
            ### Display ###
            print(f"Healing by {self.get_name()} on {patient.get_name()} is unsuccessful")
            print(f"    Current Strength {self.get_strength()} less than required strength {r}")
            
        elif self.get_strength() >= r: # Enough strength to attack
            ### Set Variables ###
            patient.health = max(100, patient.health + r)
            self.strength -= r # Decrease strength by r
            
            ### Display ###
            print(f"Healing by {self.get_name()} on {patient.get_name()} is successful")
            print(f"    Current Strength dropped by {r} to {self.get_strength()} ")
            print(f"    Patient {patient.get_name()} health rose by {r} to {patient.health}")
```

```python
# Task 1.4

## Define the class Warrior
class Warrior(Person):
    def attack(self, opponent):
        r = random.randint(1, 10) # Only new code is here
        ### From above ###########################################################
        if self.get_strength() < r: # Not enough strength to attack
            ### Display ###
            print(f"Attack by {self.get_name()} on {opponent.get_name()} is unsuccessful")
            print(f"    Current Strength {self.get_strength()} less than required strength {r}")
            
        elif self.get_strength() >= r: # Enough strength to attack
            ### Set Variables ###
            opponent.health -= r * 2
            self.strength -= r # Decrease strength by r
            
            ### Display ###
            print(f"Attack by {self.get_name()} on {opponent.get_name()} is successful")
            print(f"    Current Strength dropped by {r * 2} to {self.get_strength()} ")
            print(f"    Opponent {opponent.get_name()} health fell by {r} to {opponent.health}")
            ### Check defeated ###
            if opponent.health <= 0:
                print(f"Opponent {opponent.get_name()} is defeated")
                
def driver():
    cheese = Food('Cheese', 10)
    sam = Warrior('Sam')
    alex = Healer('Alex')
    jan = Person('Jan')
    jan.attack(sam)
    sam.attack(jan)
    alex.heal(jan)
    sam.eat(cheese)
driver()
```
