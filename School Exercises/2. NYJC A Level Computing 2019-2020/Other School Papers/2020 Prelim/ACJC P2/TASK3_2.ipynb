```python
# Task 3.2
### From Task 3.1 ###################################################
import csv

def readFile(filename="INVENTORY.txt"):
    data = []
    with open(filename, 'r') as infile:
        reader = csv.reader(infile, delimiter='\t')
        for record in reader:
            newRecord = record
            newRecord[2] = float(record[2])
            newRecord[3] = float(record[3])
            newRecord[4] = int(record[4])
            data.append(newRecord)
    
    return data

### New Code ##########################################################
def calculateTotalProfit(data):
    profit = 0
    for record in data:
        localRevenue = record[3] * record[4] # Selling Price * Quantity
        localCost = record[2] * record[4] # Purchase Price * Quantity
        localProfit = localRevenue - localCost
        profit += localProfit
    return profit
        
def generateProductMargins(data):
    productMarginsList = []
    for record in data:
        name = record[0]
        Type = record[1]
        purchasePrice = record[2]
        sellingPrice = record[3]
        
        margin = sellingPrice - purchasePrice
        newRecord = (name, Type, margin)
        productMarginsList.append(newRecord)
    return productMarginsList

### Merge Sort ####################################
def isDescendingProfitMargin(a, b):
    return a[2] > b[2] # Descending order of profit margin

def merge(arrLeft, arrRight, inOrder):
    mergedArr = []
    left = 0
    right = 0
    while left < len(arrLeft) or right < len(arrRight):
        if ( # Left element in order to right element , or rightArr no more elements
            ( left < len(arrLeft) and right < len(arrRight) and inOrder(arrLeft[left], arrRight[right]) ) 
            or (right >= len(arrRight))
           ):
            mergedArr.append(arrLeft[left])
            left += 1
        else:
            mergedArr.append(arrRight[right])
            right += 1
    return mergedArr

#merge([1,3,5,7], [2,4,6,8], lambda a, b: a < b) # Tester

def mergeSort(arr, inOrder):
    ### Base Case ###
    if len(arr) <= 1:
        return arr
    
    ### Recursive Case ###
    mid = len(arr) // 2
    arrLeft = arr[:mid]
    arrRight = arr[mid:]
    
    # Recursive step
    arrLeft = mergeSort(arrLeft, inOrder)
    arrRight = mergeSort(arrRight, inOrder)
    
    # Merging
    return merge(arrLeft, arrRight, inOrder)

### Display ##########################################
def displayInventoryData(data):
    output = f"{'Product':20} {'Product Type':20} {'Profit Margin':20}"
    for record in data:
        output += f"\n{record[0]:20} {record[1]:20} {int(record[2]):<20}"
    print(output)
    
### Main #################################################################
def task3_2():
    data = readFile("INVENTORY.txt")
    print(f"Total profit is ${calculateTotalProfit(data):.2f}")
    print()
    profitMargins = generateProductMargins(data)
    sortedProfitMargins = mergeSort(profitMargins, isDescendingProfitMargin)
    displayInventoryData(sortedProfitMargins)
    
task3_2()
```
