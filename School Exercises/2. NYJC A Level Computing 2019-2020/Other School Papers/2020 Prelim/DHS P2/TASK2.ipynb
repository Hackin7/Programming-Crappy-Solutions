```python
## Task 2
def readFile(filename="LASTNAMES.txt"):
    # Initialisation
    allLists = []
    file = open(filename, 'r') # Read mode
    
    # Main input loop
    n = int(file.readline().strip())
    while n != 0:
        currList = []
        for i in range(n):
            name = file.readline().strip()
            currList.append(name)
        allLists.append(currList)
        
        n = int(file.readline().strip())

    # Finalisation
    file.close()
    return allLists

# Custom sort comparator
def inOrder(a, b):
    # Comparing by ASCII order
    
    ### First character #######################
    valA = a[0]
    valB = b[0]
    if valA < valB:
        return True
    elif valA > valB:
        return False
    elif valA == valB:
        ### Second character #######################
        valA = a[1]
        valB = b[1]
        if valA < valB:
            return True
        elif valA > valB:
            return False
        elif valA == valB: # Don't touch if first 2 letters are the same
            return True

def bubbleSort(array, inOrder=inOrder):
    isSorted = False
    while not isSorted:
        isSorted = True
        for i in range(0, len(array) - 1): # 1 pass through array
            if not inOrder(array[i], array[i + 1]): # Swapping
                temp = array[i]
                array[i] = array[i + 1]
                array[i + 1] = temp
                isSorted = False
        
def outputLists(allLists):
    output = ""
    for index in range(len(allLists)):
        currList = allLists[index]
        for item in currList:
            output += f"\n{item}"
            
        # Newline after each lists
        if index < len(allLists) - 1: # Remove only if not last element
            output += "\n"
    
    # Displaying
    output = output.strip()
    print(output)
            
def task2():
    allLists = readFile()
    for index in range(len(allLists)):
        bubbleSort(allLists[index])
    outputLists(allLists)
    
task2()
#bubbleSort([4,3,2,1], lambda x,y: x<y)
```
