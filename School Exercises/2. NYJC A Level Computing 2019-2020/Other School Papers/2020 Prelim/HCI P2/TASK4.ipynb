```python
# Task 4.1
# Write code for the superclass LinkedStructure

# Auxilary class Node
class Node:
    def __init__(self, data=None, next=None):
        self.data = data
        self.next = next

class LinkedStructure:
    def __init__(self):
        self.Initialise()
    
    def Initialise(self):
        # Use a python list to simulate an array
        self.size = 10
        self.array = []
        for i in range(self.size + 1): # 1-indexed. 0-indexed goes unused
            next = (i + 1) % (self.size + 1)
            node = Node(None, next)
            self.array.append(node)
        
        # Others
        self.Head = 0 # Pointer to First element in linked list
        self.Tail = 1 # Pointer to Last element in linked list
        self.Free = 1 # Pointer to First element in free list
        
        #self.array[self.Head].next = 0
    
    ### The Add and Remove Functions give me soo mucch issues! ################################################3
    def Add(self, item):
        if not self.IsFull():
            # Traverse Linked list for appropriate position
            prev = None
            curr = self.array[self.Head]
            while curr.data != None and curr.data < item:
                #print(curr.data, curr.next)
                prev = curr
                curr = self.array[curr.next]
                
            
            # ---[prev]---[new]---[curr]---
            
            # Creating a new node
            newPos = self.Free
            new = self.array[newPos]
            new.data = item
            self.Free = new.next
            
            if prev == None:# At the head
                currPos = self.Head
                new.next = currPos
                self.Head = newPos
            else: # After a node
                currPos = prev.next
                prev.next = newPos # Join previous node
                new.next = currPos # Join to next node
                
            # Insert at end
            if curr.data == None:
                self.Tail = newPos
                new.next = 0
            
        else:
            raise Exception(f"No Space left to add item {item}")
    
    def Remove(self, item):
        # Traverse Linked list for appropriate item
        prevPos = self.Head
        prev = None
        curr = self.array[self.Head]
        while curr.data != None:
            # Removal 
            if curr.data == item:
                ## Inserting in between
                if prev != None:
                    currPos = prev.next
                    prev.next = curr.next # Bypassing
                    ## Inserting at the end
                    if currPos == self.Tail:
                        self.Tail = prevPos
                ## Inserting at the front
                else:
                    currPos = self.Head
                    self.Head = curr.next
                    
                # Adding current node back to free list
                curr.next = self.Free
                curr.data = None # May as well remove data
                self.Free = currPos
                return True
            
            if prev != None:
                prevPos = prev.next
            prev = curr
            curr = self.array[curr.next]
        return False
    
    ###########################################################################################################
    
    def Display(self):
        output = "Items in order:\n"
        # Traverse Linked list
        prev = None
        curr = self.array[self.Head]
        while curr.data != None:
            output += f"{curr.data}\n"
            prev = curr
            curr = self.array[curr.next]
        print(output[:-1])
    
    def PrintStructure(self):
        """
        Displays current state of all pointers and the array contents in indexed order
        """
        
        output = ""
        
        # Pointers
        output += f"Head Pointer = {self.Head}\n"
        output += f"Tail Pointer = {self.Tail}\n"
        output += f"Free Pointer = {self.Free}\n"
        #output += "\n"
        
        # Array Contents
        output += f"{'Index':5} {'Content':20} {'Next':4}\n"
        for index in range(1,len(self.array)):
            content = str(self.array[index].data)
            nextPointer = self.array[index].next
            output += f"{index:5} {content:20} {nextPointer:4}" + "\n"
        print(output[:-1])
    
    def IsEmpty(self):
        '''
        # Check for used nodes
        for index in range(1,len(self.array)):
            content = self.array[index].data
            if content != None:
                return False
        return True
        '''
        # Another possible method is to check if self.Free = self.Head? not sure need more testing
        return self.array[self.Head].data == None
        
    def IsFull(self):
        # Check for unused nodes
        for index in range(1,len(self.array)):
            content = self.array[index].data
            if content == None:
                return False
        return True
```

```python
# Task 4.3
class Queue(LinkedStructure):
    def Add(self, item):
        ## Set new node with item
        newPos = self.Free
        new = self.array[newPos]
        new.data = item
        self.Free = new.next
        new.next = 0 #Cap off the linked list
        
        ## Insert from the tail
        tail = self.array[self.Tail]
        tail.next = newPos
        self.Tail = newPos
    
    def Remove(self):
        # Remove from the head
        if not self.IsEmpty():
            currPos = self.Head
            head = self.array[self.Head]
            # Move the Head position
            self.Head = head.next
            # Add back to free list
            head.next = self.Free
            data = head.data
            head.data = None
            self.Free = currPos
            return data
        else:
            return None
        
    def Display(self):
        output = "Items from earliest inserted to last inserted:\n"
        # Traverse Linked list
        prev = None
        curr = self.array[self.Head]
        while curr.data != None:
            output += f"{curr.data}\n"
            prev = curr
            curr = self.array[curr.next]
        print(output[:-1])
```

```python
# Task 4.4
def task4_3():
    # Create a new Queue
    queue = Queue()
    # Add data into queue
    file = open('QUEUE.TXT', 'r')
    for line in file:
        data = line.strip()
        print("Inserting", data)
        queue.Add(data)
    file.close()
    
    # Output current contents of queue
    print('\n')
    queue.Display()
    
    # Remove and Output 2 items from the queue
    print('\n')
    removedItems = []
    for i in range(2):
        removedItems.append(queue.Remove())
    print(f"Removed {removedItems}")
    
    # Output all pointers and array contents of the queue after the removal of the items
    print('\n')
    queue.PrintStructure()
    
task4_3()
```
