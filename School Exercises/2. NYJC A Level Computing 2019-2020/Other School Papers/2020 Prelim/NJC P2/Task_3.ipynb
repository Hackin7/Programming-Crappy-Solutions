```python
#Task 3.5
class Bed:
    def __init__(self, floor, ward, bedNo, visitHourStart, visitHourEnd):
        self.floor = int(floor)
        self.ward = int(ward)
        self.bedNo = int(bedNo)
        self.occupiedBy = None # Patient
        self.queue = Queue()
        self.visitHourStart = int(visitHourStart)
        self.visitHourEnd = int(visitHourEnd)
        
    def __str__(self):
        name = None
        if self.occupiedBy:
            name = self.occupiedBy.getName()
        output =  f"{self.floor}-{self.ward}-{self.bedNo}:{name}"
        return output
```

```python
#Task 3.6
class Hospital:
    def __init__(self, noFloors, noWards, noBeds):
        self.noFloors = int(noFloors)
        self.noWards = int(noWards)
        self.noBeds = int(noBeds)
        # Python list simulate an array of beds
        self.beds = []
        for floor in range(1, self.noFloors + 1):
            ## Hours
            if floor == self.noFloors: # Highest floor
                startHour = 17
                endHour = 19
            else:
                startHour = 12
                endHour = 20
            # Wards and more
            for ward in range(1, self.noWards+1):
                for bedNo in range(1, self.noBeds+1):
                    self.beds.append(Bed(floor, ward, bedNo, startHour, endHour))
        
    def visit(self, visitor, floor, ward, bedNo, timestamp):
        floor = int(floor)
        ward = int(ward)
        bedNo = int(bedNo)
        index = self.hash(floor, ward, bedNo)
        bed = self.beds[index]
        
        if bed.occupiedBy == None:
            return False
        elif bed.visitHourStart < timestamp.hour < bed.visitHourEnd:
            self.beds[index].queue.enqueue(visitor)
            return True
        else:
            return False
    
    def hash(self, floor, ward, bedNo):
        floor = int(floor)
        ward = int(ward)
        bedNo = int(bedNo)
        return (bedNo - 1) + (ward - 1) * self.noBeds + (floor - 1) * self.noWards
    
    def occupy(self, patient, floor, ward, bedNo):
        floor = int(floor)
        ward = int(ward)
        bedNo = int(bedNo)
        index = self.hash(floor, ward, bedNo)
        self.beds[index].occupiedBy = patient
        #print(patient, floor,ward, bedNo, index)
        #print(self.beds[index])
        
    def showOccupancy(self):
        for index in range(len(self.beds)):
            if self.beds[index].occupiedBy != None:
                print(f"{index}->{str(self.beds[index])}{str(self.beds[index].queue)}")
```
