```python
#Task 2.1

import json
def readFile(filename="RESTAURANTS.JSON"):
    file = open(filename, 'r') # Read mode
    strdata = file.read()
    file.close()
    data = json.loads(strdata) # Convert to python lists and dictionaries
    
    #print(len(data))
    ### Remove duplicates ##############################################
    processedData = []
    backlog = [] # list of restaurant names
    for item in data:
        if item['name'] not in backlog: # If it is not a duplicate
            ### Processing item ################################
            processedItem = item
            # Adding rating
            if processedItem.get('rating') == None or \
                not str(processedItem.get('rating')).isdigit(): 
                processedItem['rating'] = 0
            else:
                processedItem['rating'] = int(processedItem['rating'])
            ### Adding to fitered data structure ##################
            backlog.append(processedItem['name'])
            processedData.append(processedItem)
    
    #print(len(set(backlog)))
    return processedData

restaurants = readFile()
```

```python
#Task 2.2
def cmp(a, b): # Returns if the item is in the correct order
    #return a < b # Testing
    ## Descending order of rating values
    return a['rating'] > b['rating']

def merge(left, right):
    new = []
    # Looping indexes
    l = 0
    r = 0
    while l < len(left) or r < len(right): # Add until elements of both lists are added
        # Add smaller item from each list
        #  or add if all items from other list is exhausted
        if r >= len(right) or (l < len(left) and cmp(left[l],right[r])): 
            new.append(left[l])
            l += 1
        else:
            new.append(right[r])
            r += 1
    return new

def mergeSort(arr):
    ### Base case: Already sorted ##########
    if len(arr) <= 1:
        return arr
    ### Other case #########################
    else:
        mid = int(len(arr)/2)
        left = arr[:mid]
        right = arr[mid:]
        ### Recursive case ################
        left = mergeSort(left)
        right = mergeSort(right)
        ### Merger
        return merge(left, right)

#merge([1,2,3],[2,3,4]) # Testing
#mergeSort([9,8,7,6,5]) # Testing

```

```python
def task2_3():
    sortedRestaurants = mergeSort(restaurants)
    print("Top 10 restaurants:")
    print(f"{'Restaurant':<20} | Rating")
    for i in range(10):
        name = sortedRestaurants[i]['name']
        print(f"{name:<20} | {sortedRestaurants[i]['rating']}")
        
task2_3()
```
