```python
# Task 2.1

class Queue:
    def __init__(self):
        # Use a python list to simulate an array
        size = 4 + 1
        self.Items = [""] * size # "" represents empty space
        # Pointers
        self.Front = int(0) # Index of first item added to the queue
        self.Rear = int(0) # Index of last item added to the queue
        
    def IsEmpty(self):
        return self.Items[self.Front] == "" and self.Rear == self.Front
    
    def IsFull(self):
        return self.Rear == len(self.Items) - 1 # At the last pointer
    
    def Enqueue(self, data: str):
        if self.IsFull():
            print(f'The queue is full, cannot insert {data}!')
        else:
            if not self.IsEmpty():
                self.Rear += 1
            self.Items[self.Rear] = data
    
    def Dequeue(self) -> str:
        if self.IsEmpty():
            print("The queue is empty, nothing to dequeue!")
            return ""
        else:
            data = self.Items[self.Front]
            self.Front += 1
            return data
        
    def Display(self):
        if self.IsEmpty():
            print("Nothing in queue to display")
        else:
            print("Items in queue:")
            # Traverse the queue
            curr = self.Front
            while curr != self.Rear:
                print(self.Items[curr])
                curr += 1
            print(self.Items[curr]) # Print the last element

class CircularQueue(Queue):
    def __init__(self):
        super().__init__()
        
    def IsFull(self):
        # Get size first
        if self.IsEmpty():
            size = 0
        elif self.Front < self.Rear:
            size = self.Rear - self.Front + 1
        elif self.Front >= self.Rear:
            size = len(self.Items) - (self.Front - self.Rear - 1)
            
        return size == len(self.Items)
    
   ### Similar to LinearQueue Methods but the index loops around #############
    def Enqueue(self, data: str):
        if self.IsFull():
            print(f'The queue is full, cannot insert {data}!')
        else:
            if not self.IsEmpty():
                self.Rear = (self.Rear + 1) % len(self.Items) # Looping index
            self.Items[self.Rear] = data
    
    def Dequeue(self) -> str:
        if self.IsEmpty():
            print("The queue is empty, nothing to dequeue!")
            return ""
        else:
            data = self.Items[self.Front]
            self.Front = (self.Front + 1) % len(self.Items) # Looping index
            return data
        
    def Display(self):
        if self.IsEmpty():
            print("Nothing in queue to display")
        else:
            print("Items in queue:")
            # Traverse the queue
            curr = self.Front
            while curr != self.Rear:
                print(self.Items[curr])
                curr = (curr + 1) % len(self.Items) # Looping index
            print(self.Items[curr]) # Print the last element
```

```python
# Task 2.2
def queueTester(queue):
    for item in ["John", "Amy", "Chetan", "Xin Xin", "Evan"]:
        queue.Enqueue(item)
        #print(queue.Front, queue.Rear)
    for i in range(2):
        print("Dequeued:", queue.Dequeue())
    queue.Enqueue("Mohan")
    queue.Display()

def task2_2():
    print("### Testing Linear Queue ###")
    linearQueue = Queue()
    queueTester(linearQueue)
    print()
    
    print("### Testing Circular Queue ###")
    circularQueue = CircularQueue()
    queueTester(circularQueue)
    
task2_2()
```
