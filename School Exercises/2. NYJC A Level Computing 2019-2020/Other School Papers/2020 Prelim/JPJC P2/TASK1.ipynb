```python
# Task 1.1

import csv
def readFile(filename="SURVEY.TXT"):
    # Initialisation 
    Records = []
    infile = open(filename, 'r')
    reader = csv.reader(infile)
    
    # Reading from file
    for record in reader:
        newRecord = {
            "Day" : record[0], 
            "A" : int(record[1]), 
            "B" : int(record[2]), 
            "C" : int(record[3]), 
            "D" : int(record[4]), 
            "E" : int(record[5]), 
            "F" : int(record[6]), 
        }
        Records.append(newRecord)
    
    infile.close()
    return Records

def displayRecords(Records):
    if len(Records) == 0:
        print("No Record to display")
        return
    
    output = ""
    keys = list(Records[0].keys())
    colwidth = [10, 10, 10, 10, 10, 10, 10]
    
    # Headings
    row = ""
    for index in range(len(keys)):
        row += f"{keys[index]:<{colwidth[index]}}"
    output += row
    
    # Data
    for record in Records:
        row = ""
        for index in range(len(keys)):
            row += f"{record[keys[index]]:<{colwidth[index]}}"
        output += "\n" + row
    
    print(output)
    
### Main Driver ####################
def task1_1():
    Records = readFile()
    displayRecords(Records)
task1_1()
```

```python
# Task 1.2

def calcLateArrivalsPerRoute(Records):
    # Initialisation
    routes = ['A', 'B', 'C', 'D', 'E', 'F']
    noLate = {}
    totalLateMin = {}
    
    
    for route in routes:
        noLate[route] = 0
        totalLateMin[route] = 0
    
    # Calculate no. late arrivals and total minutes late
    for record in Records:
        for route in routes:
            if record[route] < 0:
                noLate[route] += 1
                totalLateMin[route] += -record[route]
    
    # Calculate average minutes late per route
    avgLateMin = {}
    for route in routes:
        if noLate[route] == 0:
            avgLateMin[route] = 0
        else:
            avgLateMin[route]= totalLateMin[route] / noLate[route]
    
    # Find bus route with highest no.days late:
    highestNoDaysLate = noLate[routes[0]]
    routesMostLate = [routes[0]]
    for route in routes[1:]: # Start from 2nd route
        if noLate[route] > highestNoDaysLate:
            highestNoDaysLate = noLate[route]
            routesMostLate = [route]
        elif noLate[route]== highestNoDaysLate:
            routesMostLate.append(route)
        
    return noLate, avgLateMin, routesMostLate

def displayCalculations(noLate, avgLateMin, routesMostLate):
    output = ""
    keys = ['', 'A', 'B', 'C', 'D', 'E', 'F']
    colwidth = [40, 10, 10, 10, 10, 10, 10]
    
    ### Table ######################
    # Headings
    row = ""
    for index in range(len(keys)):
        row += f"{keys[index]:<{colwidth[index]}}"
    output += row
    
    # Data
    data = [ ['No. of late arrivals for each bus route'], ['Average no. of minutes late']]
    for route in noLate.keys():
        data[0].append(noLate[route])
        data[1].append(round(avgLateMin[route],1))
    
    for record in data:
        row = ""
        for index in range(len(keys)):
            # Formatting (1 decimal place)
            data = record[index]
            if record[0] == 'Average no. of minutes late' and (type(data) == float or type(data) == int):
                formatting = '.1f'
            else:
                formatting = ''
            # Row
            row += f"{data:<{colwidth[index]}{formatting}}"
        output += "\n" + row
    
    ### Highest number of days late ###########
    routeString = ""
    for route in routesMostLate:
        routeString += f"{route}, "
    routeString = routeString[:-2]
    
    output += f'\nBus route {routeString} has the highest number of days late'
    print(output)
    
### Main Driver ####################
def task1_2():
    Records = readFile()
    displayRecords(Records)
    print()
    noLate, avgLateMin, routesMostLate = calcLateArrivalsPerRoute(Records)
    displayCalculations(noLate, avgLateMin, routesMostLate)
task1_2()
```

```python
# Task 1.3

def task1_3():
    day = input("Enter a specific day: ").strip()
    Records = readFile()
    
    foundIndex = -1
    for index in range(len(Records)):
        if Records[index]['Day'] == day:
            foundIndex = index
    if foundIndex == -1:
        print("The Day is not found")
        return
    
    output = ""
    count = 0
    routes = ['A', 'B', 'C', 'D', 'E', 'F']
    for route in routes:
        if Records[foundIndex][route] < 0: # is Late
            count += 1
            output += f"\nThe day is late on route {route} by {-Records[foundIndex][route]} minutes"
            
    output = f"The day has {count} late buses"+ output
    print(output)
    
task1_3()
task1_3()
```
