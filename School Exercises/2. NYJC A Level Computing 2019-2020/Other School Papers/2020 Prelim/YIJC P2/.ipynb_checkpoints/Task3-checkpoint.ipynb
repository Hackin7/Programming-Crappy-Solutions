```python
# Task 3.1
class Node:
    def __init__(self):
        self.set_word('')
        self.set_category('')
        self.set_next(0)
    
    ### Setter Methods ##########################
    def set_word(self, w: str):
        self.__word = str(w)
    
    def set_category(self, c: str):
        self.__category = str(c)
    
    def set_next(self, n: int):
        self.__next = int(n)
    
    ### Getter Methods #########################
    def get_word(self) -> str:
        return self.__word
    
    def get_category(self) -> str:
        return self.__category
    
    def get_next(self) -> int:
        return self.__next

class LinkedList:
    def __init__(self):
        # Use Python list to simulate Array
        size = 10
        self.sentence = [] 
        for i in range(size): # 0-indexed
            # Initialising values
            nextPos = (i + 1) % size
            if nextPos == 0:
                nextPos = -1 # Terminating condition
            # New node
            newNode = Node()
            newNode.set_next(nextPos)
            self.sentence.append(newNode)
            
        # Pointers
        self.start = -1
        self.nextfree = 0
        
    def isempty(self):
        return self.start == -1
    
    def isfull(self):
        return self.nextfree == -1
    
    def display(self):
        index, word, category, nextPos = "Index", "Word", "Category", "Next"
        output = f"{index:<7} {word:<10} {category:<10} {nextPos:<5}"
        for index in range(0, len(self.sentence)):
            index, word, category, nextPos = (
                index,
                self.sentence[index].get_word(), 
                self.sentence[index].get_category(), 
                self.sentence[index].get_next()
            )
            output += f"\n{index:<7} {word:<10} {category:<10} {nextPos:<5}"
        
        print(output)
```

```python
# Task 3.2
def task3_2():
    ll = LinkedList()
    ll.display()
    return ll
ll = task3_2()
```

```python
# Task 3.3
class LinkedList(LinkedList):
    def insert(self, new_word, new_category):
        # check if Linked List is full
        if self.nextfree == -1:
            return 'Linked List is Full'
        
        # linked list is not full, safe to insert new word
        self.sentence[self.nextfree].set_word(new_word)
        self.sentence[self.nextfree].set_category(new_category)

        if self.start == -1:      # inserting into empty list
            self.start = self.nextfree
            temp = self.sentence[self.start].get_next()
            self.sentence[self.start].set_next(-1)
            self.nextfree = temp                    


        else:      #  traverse down the linked list to search for position to insert
            current = self.start  # pointer of current node
            previous = -1	# pointer of previous node
            inserted = False # flag to check for insertion

            while current > -1 and inserted == False:
                if self.sentence[current].get_category() > new_category: 	  
                # position found, insert before current node   
                    if current == self.start: 
                        # check if current equals to self.start
                        self.start = self.nextfree
                    else:
                        self.sentence[previous].set_next(self.nextfree)


                    temp = self.sentence[self.nextfree].get_next()
                    self.sentence[self.nextfree].set_next(current)
                    self.nextfree = temp 
                    inserted = True 


                elif self.sentence[current].get_category() < new_category: 
                    previous = current
                    current = self.sentence[current].get_next()

                else: 
                    if new_category == 'N':
                        new_category = 'Z'

                    if new_category == 'D':
                        new_category = 'X'

                    if new_category == 'J': 
                        new_category = 'Y'


                    previous = current
                    current = self.sentence[current].get_next()
                    self.sentence[self.nextfree].set_category(new_category)

            if inserted == False:
                self.sentence[previous].set_next(self.nextfree)
                temp = self.sentence[self.nextfree].get_next()
                self.sentence[self.nextfree].set_next(-1)
                self.nextfree = temp 		

```

```python
# Task 3.4

data = [
    ['little', 'J'],
    ['tough', 'J'],
    ['boy', 'N'],
    ['The', 'D'],
    ['runs', 'V'],
    ['race', 'N'],
    ['a', 'D']
]

def task3_4():
    ll = LinkedList()
    for new_word, new_category in data:
        ll.insert(new_word, new_category)
    ll.display()
task3_4()
```

```python
# Task 3.5 
class LinkedList(LinkedList):
    def traverse(self):
        if self.isempty():
            print('Nothing in the Linked List')
        else:
            print('Traversing the Linked List gives you:')
            output = ""
            curr = self.start
            while curr != -1:
                node = self.sentence[curr]
                output += f"{node.get_word()} "
                curr = node.get_next()
                
            output = output.strip()
            print(output + '.')

def task3_5():
    ll = LinkedList()
    for new_word, new_category in data:
        ll.insert(new_word, new_category)
    ll.traverse()
task3_5()
```
