```python
# Task 2.1
def readWords(filename='SONG.TXT'):
    words = []
    infile = open(filename, 'r')
    for line in infile:
        wordLine = line.strip().split(' ')
        words = words + wordLine
    infile.close()
    return words

### Quick Sort ##################################################
def inOrder(a, b):
    return a <= b

def quickSort(arr, low, high, inOrder):
    # Base Case: Already Sorted
    if low >= high:
        return 
    
    # Other case
    pivot = low
    left = low + 1
    right = high
    while left <= right:
        # Traversing to positions
        while left <= right and inOrder(arr[left],arr[pivot]):
            left += 1
        while left <= right and inOrder(arr[pivot],arr[right]):
            right -= 1
        # Swapping
        if left <= right:
            temp = arr[left]
            arr[left] = arr[right]
            arr[right] = temp
    
    # Swapping in pivot
    temp = arr[pivot]
    arr[pivot] = arr[right]
    arr[right] = temp
    quickSort(arr, low,right - 1, inOrder)
    quickSort(arr, right + 1, high, inOrder)

#arr = [0, 1, 2, 2, 3];quickSort(arr, 0, len(arr) - 1, lambda a, b: a >= b);print(arr) # Debugging

### Counting ##################################################
def count(arr):
    newData = []
    prevWord = arr[0]
    prevCount = 0 # Add at first
    for word in arr:
        if word == prevWord:
            prevCount += 1
        else:
            newData.append((prevWord, prevCount))
            prevWord = word
            prevCount = 1
    return newData

def outputCountedData(countedData):
    output = ""
    for index in range(len(countedData)):
        record = countedData[index]
        output += f"{record[0]} {record[1]}\n"
    return output
        
### Main #########################################
def task2_1():
    words = readWords()
    quickSort(words, 0, len(words) - 1, inOrder)
    wordsCounted = count(words)
    
    # Write
    outfile = open('WORDCOUNT.TXT', 'w')
    outfile.write(outputCountedData(wordsCounted))
    outfile.close()

task2_1()
```

```python
# Task 2.2
import csv
def readWordCount(filename="WORDCOUNT.TXT"):
    wordToCount = {}
    countToWord = {}
    infile = open(filename, 'r') # Read Mode
    reader = csv.reader(infile, delimiter=' ')
    for record in reader:
        word = record[0]
        count = int(record[1])
        
        wordToCount[word] = count
        if countToWord.get(count) == None:
            countToWord[count] = [word]
        else:
            countToWord[count].append(word)
        
    infile.close()
    return wordToCount, countToWord
    
def menu(wordToCount, countToWord):
    print('1. Search for a word')
    print('2. Search for word(s) based on count')
    print('3. Quit program')
    print()
    option = "0"
    while option != "3":
        option = input("Enter option: ").strip()
        
        ##### Options ########################################
        ### Get count from word
        if option == "1": 
            word = input("Enter word: ").strip()
            count = wordToCount.get(word)
            if count == None:
                print("Not Found")
            else:
                print(f'The word {word} is counted {count} times')
        ### Get word from count
        elif option == "2": 
            count = int(input("Enter count: "))
            
            words = countToWord.get(count)
            if words == None:
                print('Not Found')
            else:
                countStr = ''
                for word in words:
                    countStr += word + ','
                countStr = countStr[:-1]
                print(f'The count {count} is for these words: {countStr}')
        ### Others
        elif option == '3':
            print('Bye')
        else:
            print('Invalid Option')
        print()
    
def task2_2():
    wordToCount, countToWord = readWordCount()
    menu(wordToCount, countToWord)
    
task2_2()
```
