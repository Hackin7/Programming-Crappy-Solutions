```python
# Task 3.1
class Node:
    def __init__(self): # constructor()
        # Simulate private
        self.__key = None
        self.__left = None
        self.__right = None 
        
    def insert(self, key:int):
        newNode = Node()
        newNode.setKey(key)
        if key < self.getKey(): # Insert left
            if self.__left == None:
                self.__left = newNode
            else:
                self.__left.insert(key)
        else: # Insert right
            if self.__right == None:
                self.__right = newNode
            else:
                self.__right.insert(key)
            
    ### Additional methods ####################
    def getKey(self):
        return self.__key
    
    def setKey(self, key:int):
        self.__key = int(key)
        
    def printInOrder(self):
        if self.__left != None:
            self.__left.printInOrder()
            
        print(self.getKey())
        
        if self.__right != None:
            self.__right.printInOrder()
        
class Tree:
    def __init__(self):
        self.__root = None
    
    ### Additional Getter Setters ################
    def getRoot(self):
        return self.__root
    
    def setRoot(self, val):
        self.__root = val
        
    def add(self, newItem):
        if self.getRoot() == None: # Empty
            self.setRoot(Node())
            self.getRoot().setKey(newItem)
        else:
            self.getRoot().insert(newItem)
            
    def printInOrder(self):
        if self.getRoot() == None: # Empty
            print("Nothing in Tree!")
        else:
            self.getRoot().printInOrder()
```

```python
# Task 3.3
def mergeSort(arr): # just for fun lmao
    ### Base Case: Already sorted ##############
    if len(arr) <= 1:
        return arr
    ### Recursive Case: sort ######################
    mid = len(arr) // 2
    left = mergeSort(arr[:mid]) # Recursive call
    right = mergeSort(arr[mid:]) # Recursive call
    ### Merging
    new = []
    while len(left) > 0 and len(right) > 0:
        if left[0] < right[0]: # left before right
            new.append(left.pop(0))
        else:
            new.append(right.pop(0))
    new = new + left + right # Add any remaining elements
    return new
    
### New Code ######################################################
def balancedInsertToTree(tree, arr, low, high):
    if low > high:
        return
    else:
        mid = (low + high) // 2
        #print(arr[mid]) # Debugging
        tree.add(arr[mid])
        balancedInsertToTree(tree, arr, low, mid - 1)
        balancedInsertToTree(tree, arr, mid + 1, high)

def CreateTreeFromArray(arr):
    tree = Tree()
    newArr = mergeSort(arr)
    #print(newArr)
    balancedInsertToTree(tree, newArr, 0, len(arr) - 1)
    tree.printInOrder()  
    #tree._Tree__root._Node__left.getKey()
      
# Copying from BST.txt
arr = [11, 6, 1, 14, 16, 7, 17, 20, 13, 9, 15, 8, 5, 4, 2]
CreateTreeFromArray(arr)
```
