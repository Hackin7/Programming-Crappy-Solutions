```python
# Task 4.1

class ListNode:
    def __init__(self, ID: str, Score: int, Ptr: int):
        self.ID = str(ID)
        self.Score = int(Score)
        self.Ptr = int(Ptr)
    
    def setID(self, ID):
        self.ID = str(ID)
        
    def setScore(self, Score):
        self.Score = int(Score)
        
    def setPtr(self, Ptr):
        self.Ptr = int(Ptr)
    
class LinkedList:
    def __init__(self, size=20):
        # Use a python list to simulate an array
        self.Node = []
        for i in range(size + 1): # 1-indexed, ignore first element
            nextPtr = (i + 1) % (size + 1) 
            self.Node.append(ListNode('', -1, nextPtr))
        
        # Other pointers    
        self.HeadPtr = 1
        self.FreePtr = 1
```

```python
# Task 4.2

class LinkedList(LinkedList):
    def traverse(self, reachedFunc):
        prevPtr = None
        currPtr = self.HeadPtr
        while currPtr != 0 and not reachedFunc(currPtr, self.Node[currPtr]):
            prevPtr = currPtr
            currPtr = self.Node[currPtr].Ptr
        return prevPtr, currPtr
    
    def insert(self, prevPtr, currPtr, ID, Score):
        #print(ID, Score, prevPtr, currPtr)
        
        # Use new node
        newPtr = self.FreePtr
        self.Node[newPtr].setID(ID)
        self.Node[newPtr].setScore(Score)
        self.FreePtr = self.Node[newPtr].Ptr # Adjust free list
        
        # Link new node
        if prevPtr == None: # Insert at start
            if self.HeadPtr == newPtr: # If the thing is empty
                self.Node[newPtr].setPtr(0) # Ending the linked List
            else:
                self.Node[newPtr].setPtr(currPtr)
            self.HeadPtr = newPtr

        else: # Insert in between
            self.Node[prevPtr].setPtr(newPtr)
            self.Node[newPtr].setPtr(currPtr)
            
    def AddInOrder(self, ID, Score):
        def reachedFunc(currPtr, currNode):
            return currNode.Score <= Score
        prevPtr, currPtr = self.traverse(reachedFunc)
        self.insert(prevPtr, currPtr, ID, Score)
```

```python
# Task 4.5
class LinkedList(LinkedList):
    def AddInOrder(self, ID, Score):
        def reachedFunc(currPtr, currNode):
            if currNode.Score < Score:
                return True
            elif currNode.Score == Score:
                # Check alphabetical order
                return currNode.ID > ID
            else:
                return False
        
        prevPtr, currPtr = self.traverse(reachedFunc)
        self.insert(prevPtr, currPtr, ID, Score)
        
def task4_5():
    ll = LinkedList()
    readFile(ll, 'SCORES2.txt')
    ll.OutputData()
task4_5()
```

```python
# Task 4.6
class LinkedList(LinkedList):
    def DisplayByRank(self):
        output = ""
        # Headings
        headers = ['Rank', 'Player ID', 'Score']
        colwidth = [10, 10, 10]
        row = ""
        for i in range(len(headers)):
            row += f"{headers[i]:<{colwidth[i]}}"
        output += row
        
        count = 0
        rank = 0
        rankScore = 100000000
        
        # Traverse Linked List
        prevPtr = None
        currPtr = self.HeadPtr
        while currPtr != 0:# and not reachedFunc(currPtr, self.Node[currPtr]):
            # Processing
            count += 1
            if self.Node[currPtr].Score < rankScore:
                rank = count
                rankScore = self.Node[currPtr].Score
            
            # Display
            row = ""
            data = [rank, self.Node[currPtr].ID, self.Node[currPtr].Score]
            for i in range(len(headers)):
                row += f"{data[i]:<{colwidth[i]}}"
            output += '\n' + row
            
            # Move on
            prevPtr = currPtr
            currPtr = self.Node[currPtr].Ptr
        
        print(output)
        
def task4_6():     
    ll = LinkedList()
    readFile(ll, 'SCORES2.txt')
    ll.DisplayByRank()

task4_6()
```

```python
# Task 4.7
class LinkedList(LinkedList):
    def ReverseTraversal(self, currPtr=None):
        if currPtr == None:
            currPtr = self.HeadPtr
            
        ### Traverse ###
        if self.Node[currPtr].Ptr != 0:
            self.ReverseTraversal(self.Node[currPtr].Ptr)
            
        ### Displaying ###
        colwidth = [10, 10]
        row = ""
        data = [self.Node[currPtr].ID, self.Node[currPtr].Score]
        for i in range(len(data)):
            row += f"{data[i]:<{colwidth[i]}}"
        print(row)
        
def task4_7():
    ll = LinkedList()
    readFile(ll, 'SCORES2.txt')
    ll.ReverseTraversal()

task4_7()
```
