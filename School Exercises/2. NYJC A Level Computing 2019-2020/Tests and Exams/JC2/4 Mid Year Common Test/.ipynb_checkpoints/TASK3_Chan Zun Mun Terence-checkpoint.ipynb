```python
#Task 3.1

### For use in a print queue
class Queue: # Python List implemenation of a queue
    def __init__(self, size=10):
        self.max_size = max(max(0,size),10)
        self.list = [] 
    
    def size(self):
        return len(self.list)
    
    def enqueue(self, item):
        if self.size() < self.max_size: # Didn't manage to implement max size in time
            self.list = [item] + self.list # Wrong enqueuing method
    
    def dequeue(self):
        return self.list.pop()
    
    def next(self):
        if len(self.list):
            return self.list[-1]
        else:
            return None
    
    def getPrintQueue(self):
        return self.list

class PrintQueue(Queue):
    def __init__(self, name, size=10):
        Queue.__init__(self, size)
        self.name = name
    
    def setName(self, name):
        self.name = name
    
    def getName(self):
        return self.name

class PrintJob:
    def __init__(self):
        self.userID = "" # "" is null value
        self.terminalNumber = 0 # To be in range 1-200, 0 is null value
        self.jobType = '-' # Either 'C' or 'B', '-' is null value
        self.fileSize = 0 # In kilobytes, 0 is null value
    
    # Setter
    def setUserID(self, userID):
        self.userID = userID
    
    def setTerminalNumber(self, terminalNumber):
        if type(terminalNumber) == int and \
            0 <= terminalNumber <= 200:
            self.terminalNumber = terminalNumber
        else:
            raise Exception("Invalid Terminal Number")
            
    def setJobType(self, jobType):
        if jobType in ['C', 'B', '-']:
            self.jobType = jobType
        else:
            raise Exception("Invalid Job Type")
            
    def setFileSize(self, fileSize):
        if type(fileSize) == int and fileSize >= 0:
            self.fileSize = fileSize
        else:
            raise Exception("Invalid File Size")
    
    def set(self, userID, terminalNumber, jobType, fileSize):
        self.setUserID(userID)
        self.setTerminalNumber(terminalNumber)
        self.setJobType(jobType)
        self.setFileSize(fileSize)
    
    # Getter
    def getUserID(self):
        return self.userID
    
    def getTerminalNumber(self):
        return self.getTerminalNumber
    
    def getJobType(self):
        return self.jobType
    
    def getFileSize(self):
        return self.fileSize
    
    def get(self):
        return (self.userID, self.terminalNumber, 
               self.jobType, self.fileSize)
    
```

```python
#Task 3.2
MENU_TEXT = '''\
1. Add print job to print queue
2. Output next print job from printer
3. Display current print queue (all jobs in queue)
4. End simulation
'''
def menu():
    printQueue = initialisePrintQueue("Comp_Lab2")
    
    option = 5
    while option != 4:
        print(MENU_TEXT)
        option = int(input("Enter option: "))
        if option == 1:
            addPrintJob(printQueue)
        elif option == 2:
            outputNextPrintJob(printQueue)
        elif option == 3:
            displayPrintQueue(printQueue)
        print() # new line
```

```python
#Task 3.3

def initialisePrintQueue(name = "Comp_Lab2"):
    size = int(input("Enter maximum print queue size from 0-10 (will cut off): "))
    return PrintQueue(name, size)

def displayPrintQueue(printQueue):
    printJobs = printQueue.getPrintQueue()
    
    if len(printJobs) == 0:
        print("No jobs in the print queue")
        return
    
    output = f"Print Jobs for {printQueue.name}: First to last\n"
    output += f"{'User ID':20} | {'Terminal Number':20} | {'Job Type':20} | File Size (in Kbytes) "+"\n"
    for job in printJobs[::-1]:
        printJob = job.get()
        #print(printJob)
        if printJob[2] == 'C':
            jobType = 'Colour'
        else:
            jobType = 'Black'
        output += f"{printJob[0]:20} | {printJob[1]: 20} | {jobType:20} | {printJob[3]:20}\n"
    
    print(output)
```

```python
#Task 3.4
def addPrintJob(printQueue):
    if printQueue.size() >= printQueue.max_size:
        print("The print queue is full!")
        return
    
    userID = input("Enter user ID: ")
    terminalNumber = 200 #int(input("Enter terminalNumber")) # Fixed terminal Number
    jobType = input("Enter C for Colour and B for Black and White: ")
    fileSize = int(input("Enter the file size:"))
    
    newJob = PrintJob()
    newJob.set(userID, terminalNumber, jobType, fileSize)
    printQueue.enqueue(newJob)
    
def outputNextPrintJob(printQueue):
    printJob = printQueue.next()
    if printJob == None:
        print("No next print job")
        return
    
    printJob = printJob.get()
    
    output = f"Next Print Job for {printQueue.name}\n"
    output += f"{'User ID':20} | {'Terminal Number':20} | {'Job Type':20} | File Size (in Kbytes) "+"\n"
    
    if printJob[2] == 'C':
        jobType = 'Colour'
    else:
        jobType = 'Black'
    output += f"{printJob[0]:20} | {printJob[1]: 20} | {jobType:20} | {printJob[3]:20}\n"
    print(output) # Forgot to put this in exam

menu()
```
