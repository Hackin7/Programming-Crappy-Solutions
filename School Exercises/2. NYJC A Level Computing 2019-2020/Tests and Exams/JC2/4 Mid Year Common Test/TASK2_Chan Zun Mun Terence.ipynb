```python
#Task 2.2

### Other Code #########################################

import random
def displayInRows(array, sets=10):
    output = ""
    for index in range(len(array)):
        output += f"{array[index]}"
        index += 1
        if index % sets == 0:
            output += '\n'
        else:
            output += ' '
        
    output = output.strip() # Remove any trailing newline/ spaces
    print(output)

## From previous merge sort ###
def cmp(order, a, b):
    if order == 'A':
        return a < b
    else:
        return a > b

def merge(left, right, order='A'):
    result = []
    l = 0 # Index of left array
    r = 0 # Index of right array
    while l + r < len(left) + len(right):
        bothIndexesInRange = l < len(left) and r < len(right)
        if (bothIndexesInRange and cmp(order, left[l],right[r])) or \
            r >= len(right):
            result.append(left[l])
            l += 1
        else:
            result.append(right[r])
            r += 1
    return result

### New Merge Sort Code ######################################
def MergeSortPass(Array, sz, SortOrder): # Working
    # Sort the smaller subarrays
    index = 0
    while index < len(Array):
        low = index
        mid = index + sz
        high = index + 2 * sz
        #print(low, mid, high, index) # Debugging

        left = Array[low:mid]
        right = Array[mid:high]

        result = merge(left, right, SortOrder)
        for index in range(len(result)):
            Array[low + index] = result[index] 

        index = high # Sort the next subarray
        
def MergesortLoop(Array, SortOrder):
    sz = 1 # Size of subarray to sort
    while sz < len(Array):
        MergeSortPass(Array, sz, SortOrder)
        sz *= 2 # Double the array size
    return Array

### New driver which allows specifying sort order #################
def driver():
    print("Choose sort order")
    SortOrder = input("Enter 'A' for Ascending, 'D' for Descending: ")
    print() 
    
    ### Create an array of 50 random integers
    array = [] # Python list simulates an array
    for i in range(50):
        array.append(random.randint(0,99)) # Inclusive
        
    print()
    print("Original array:")
    displayInRows(array)
    
    print()
    print("Sorted array:")
    sortedArray = MergesortLoop(array, SortOrder)  
    displayInRows(sortedArray)
    
driver()
```
