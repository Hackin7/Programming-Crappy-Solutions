```python
## Task 3.1

class node:
    def __init__(self, value, prev=None, next=None):
        self.prev = prev
        self.value = value
        self.next = next

class SongPlaylist:
    def __init__(self, name):
        self.head = None
        self.name = name
        
    def insert(self, title: str):
        title = str(title)
        if self.head == None:
            self.head = node(title)
        else:
            new = node(title, next=self.head)
            self.head.prev = new
            self.head = new
        
    def insertafter(self, searchtitle: str, newtitle: str):
        searchtitle = str(searchtitle)
        newtitle = str(newtitle)
        if self.head == None:
            self.head = node(newtitle)
        else:
            # Traverse the linked list
            prev = None
            curr = self.head
            while curr != None and curr.value != searchtitle:
                prev = curr
                curr = curr.next
            # Traverse past that node
            if curr != None:
                prev = curr
                curr = curr.next
            # Insertion
            new = node(newtitle, prev=prev, next=curr)
            if prev == None: # link back
                self.head = new
            else:
                prev.next = new 
            if curr != None:
                curr.prev = new # Link forward
        
    def display(self):
        # Traverse the linked list
            prev = None
            curr = self.head
            while curr != None:
                print(curr.value)
                prev = curr
                curr = curr.next
```

```python
## Task 3.2

MENU_TEXT = '''\
1. Create New Song Playlist
2. Add a song in front
3. Add a song after
4. Display Playlist
5. -1 to End
'''

def validatePlaylist(playlists):
    playlistName = input("Enter name: ")
    while playlists.get(playlistName) == None:
        print("Invalid playlist name")
        playlistName = input("Enter name: ")
    return playlistName    
    
def menu(text=MENU_TEXT):
    playlists = {} # Corrections: Didn't put dictionary in exam
    # Menu
    print(text)
    option = "0" # Invalid option
    while option != "-1":
        option = input("Enter option: ")
        try:
            if option == "1":
                name = input("Enter name: ")
                playlists[name] = SongPlaylist(name)
            elif option == "2":
                name = validatePlaylist(playlists)
                title = input("Enter title: ")
                playlists[name].insert(title)
            elif option == "3":
                name = validatePlaylist(playlists)
                searchtitle = input("Enter existing Title:")
                newtitle = input("Enter new title: ")
                playlists[name].insertafter(searchtitle, newtitle)
            elif option == "4":
                name = validatePlaylist(playlists)
                playlists[name].display()
            elif option == "-1":
                print("Bye")
            else:
                print("Invalid Option")
        except Exception as e:
            print(e)
menu()
```

```python
## Task 3.3

def insertSorted(self, newtitle): 
    if self.head == None:
        self.head = node(newtitle)
    else:
        # Traverse the linked list
        prev = None
        curr = self.head
        while curr != None and curr.value < newtitle: # Ascending order
            prev = curr
            curr = curr.next
        # Insertion
        new = node(newtitle, prev=prev, next=curr)
        if prev == None:
            self.head = new
        else:
            prev.next = new # link back
        if curr != None:
            curr.prev = new # Link forward
    
def insertionSort(Playlist):
    sorted = SongPlaylist('sorted')
    
    # Traverse the given playlist
    prev = None
    curr = Playlist.head
    while curr != None:
        # Insert curr
        print(f'Inserting {curr.value}')
        insertSorted(sorted, curr.value)
        # Move on
        prev = curr
        curr = curr.next

    Playlist.head = sorted.head

def tester():
    playlist = SongPlaylist("MJ")
    playlist.insert("Heal the World")
    playlist.insert("Thriller")
    playlist.insert("Beat It")
    print('Inserted')
    insertionSort(playlist)
    print()
    print('Sorted Playlist')
    playlist.display()


tester()
```
