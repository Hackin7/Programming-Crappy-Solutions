```python
#Task 2.1

###Input############################
file = open("TESTDATA.txt")
data = [] #Store data from the file
for line in file:
    data.append(line[:-1].split(",")) #Slicing to remove newline character
    #data[-1] = int(data[-1])
    
file.close()

####Quick Sort############################################################
def cmp(a,b): 
    #Ascending order of filename
    return a[0]<=b[0] 

def partition(arr,low,upper):
    pivot = low
    left = low + 1
    right = upper
    while left <= right:
        #Increment pointers until they do not fufil condition
        while cmp(arr[left],arr[pivot]) and left <= right and left <= upper:
            left += 1
        while cmp(arr[pivot],arr[right]) and left <= right:
            right -= 1
        #Swapping of values
        if left <= right:
            temp = arr[right]
            arr[right] = arr[left]
            arr[left] = temp
            print(arr,left,right, arr[pivot])
    #Swapping the pivot value
    #print(arr,left,right)
    temp = arr[right]
    arr[right] = arr[pivot] 
    arr[pivot] = temp #Swap in the lower boundary
    return right # The splitpoint

def quickSort(arr, low = 0, up = -1):
    if up < 0: #Invalid value
        up = len(arr) - 1
    #Base Case
    if (up - low) <= 1:
        return
    else: #Need Sorting
        splitPoint = partition(arr,low,up)
        #print(low,up,splitPoint)
        quickSort(arr,low,splitPoint)
        quickSort(arr,splitPoint+1,up)
###Output##############################################################
def display():
    quickSort(data)
    print (f"{'Filename':20}{'Date':20}{'Size in KB':20}")
    for i in data:
        print (f"{i[0]:20}{i[1]:20}{i[2]:20}")
###Output##############################
file = open("TESTDATA_Filename_order_Chan Zun Mun Terence.txt","w")
for line in data:
    output=""
    for item in line:
        output += str(item) + ","
    output = output[:-1] + "\n"
    file.write(output)
file.close()
```

```python
#Task 2.2

#Sorting Comparator
def cmp(a,b): 
    #Ascending order of date
    return a[1]<=b[1] 
def bubbleSort(arr):
    lower = 0
    upper = len(arr)-1
    for j in range(lower, upper):
        for i in range(lower, upper):
            if not(cmp(arr[i],arr[i+1])): #If not sorted correctly
                #Swap the Values
                temp = arr[i]
                arr[i] = arr[i+1]
                arr[i+1] = temp
                #Do not terminate early
                
def display():
    bubbleSort(data)
    print (f"{'Filename':20}{'Date':20}{'Size in KB':20}")
    for i in data:
        print (f"{i[0]:20}{i[1]:20}{i[2]:20}")
    
def searchName():
    name = input("Enter the name:")
    for i in data:
        if i[0] == name:
            print("Date:",i[1])
            print("File Size (in KB)",i[2])
            return
    print("File Not Found")
    
def searchDate():
    date = input("Enter the date of the files (YYYY-MM-DD):")
    files = 0
    for i in data:
        if i[1] == date:
            print("Name:",i[0])
            print("File Size (in KB)",i[2])
            files += 1
    if files == 0:
        print("Files Not Found")
```
