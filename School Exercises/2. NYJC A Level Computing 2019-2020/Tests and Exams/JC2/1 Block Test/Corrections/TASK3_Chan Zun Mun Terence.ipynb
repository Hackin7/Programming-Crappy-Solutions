```python
#Task 3.1

class Data:
    def __init__(self, data=None):
        self.next = None
        self.past = None
        self.data = data
class DataStructure:
    def __init__(self):
        self.head = None
        self.tail = self.head #No other elements
    def display(self):
        curr = self.head
        while curr != self.tail:
            print(curr.data)
            curr = curr.next
        print(curr.data) #Print for the last element
            
```

```python
#Task 3.2

class Stack(DataStructure):
    def insert(self,value):
        newData = Data()
        newData.data = value
        # Add at the end of the list
        newData.past = self.tail
        self.tail = newData
    
    def delete(self):
        # Remove the link
        if self.tail != None: # If have elements
            self.tail = self.tail.past
            #$$$ Get and Return Value $$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$
            if self.tail!=None:
                return self.tail.data
            else:
                return None            
        #$$$ Return Appropriate value if None $$$$$$$$$$$$$$$$$$$$$$$$$$$
        return None
        
    def display(self):
        curr = self.tail # Last element (Earliest added)first
        while curr != self.head and curr != None:
            print(curr.data)
            curr = curr.past
        if curr != None:
            print(curr.data) #Print for the last element
```

```python
#Task 3.3


class Queue(DataStructure):
    def insert(self,value):
        newData = Data()
        newData.data = value
        # Add at the front of the list
        newData.next = self.head
        if self.head != None:
            self.head.past = newData
        self.head = newData
        if self.tail == None: #No Elements
            #print(self.tail)
            curr = self.head
            prev = curr
            while curr != None:
                prev = curr
                curr = curr.next
            self.tail = prev
    
    def delete(self):
        # Remove the link
        if self.tail != None: # If have elements
            self.tail = self.tail.past
        
    def display(self):
        curr = self.tail # Last element (Earliest added)first
        while curr != self.head and curr != None:
            print(curr.data)
            curr = curr.past
        if curr != None:
            print(curr.data) #Print for the last element
```

```python
#Task 3.4

###Inserting Data#####################################
task3Stack = open("TASK3stack.txt")
stack = Stack()
for line in task3Stack:
    stack.insert(int(line))
task3Stack.close()

task3Queue = open("TASK3queue.txt")
queue = Queue()
for line in task3Queue:
    queue.insert(int(line))
task3Queue.close()

###Output current contents#############################
print("Items in Stack")
stack.display()

print('\nItems in Queue')
queue.display()

###Remove items#######################################
print("\nRemoval of 2 items from stack")
for i in range(2):
    print("Item:",stack.tail.data)
    stack.delete()

print("\nRemoval of 2 items from queue")
for i in range(2):
    print("Item:",queue.tail.data)
    queue.delete()
###Output new contents#################################
print("\nItems in Stack")
stack.display()

print('\nItems in Queue')
queue.display()
```
