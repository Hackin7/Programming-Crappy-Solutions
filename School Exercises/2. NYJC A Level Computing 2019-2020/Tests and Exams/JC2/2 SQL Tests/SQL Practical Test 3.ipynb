```
import sqlite3
con = sqlite3.connect("test.db")
try:
    con.executescript("DROP TABLE Swimmers;DROP TABLE Events;DROP TABLE Results;")
    print("Cleared All past data")
except Exception as e:
    if "no such table:" not in str(e):
        print("Error:",e)

createSwimmers = "CREATE TABLE Swimmers (SwimmerID INTEGER PRIMARY KEY AUTOINCREMENT, name TEXT,gender TEXT,`date of birth` TEXT);"
createEvents = "CREATE TABLE Events (EventID INTEGER PRIMARY KEY AUTOINCREMENT, date TEXT,`stroke type` TEXT, gender TEXT,distance NUM);"
createResults = "CREATE TABLE Results (ResultsID INTEGER PRIMARY KEY AUTOINCREMENT, SwimmerID  INTEGER, EventID INTEGER,time NUM, FOREIGN KEY (SwimmerID) REFERENCES Swimmers(SwimmerID), FOREIGN KEY (EventID) REFERENCES Events(EventID));"
con.executescript(createSwimmers+createEvents+createResults)
```

```
swimmerText = """\
Swimmer: John, Male, 14-March-1990
Swimmer: Michael, Male, 12-June-1991
Swimmer: Vincent, Male, 28-September-1989
Swimmer: Katy, Female, 2-February-1990
Swimmer: Priscilla, Female, 3-March-1991
Swimmer: Rebecca, Female, 8-December-1992\
"""
data = swimmerText.split('\n')
for i in data:
    thing = i.split(", ")
    thing[0] = thing[0][9:]
    statement = "INSERT INTO Swimmers(name, gender,`date of birth`) VALUES(?,?,?);"
    con.execute(statement,thing)

eventsText = """\
Event: 10-Feb-2018, Front Crawl, Male, 100
Event: 11-Feb-2018, Front Crawl, Female, 100
Event: 12-Feb-2018, Breast Stroke, Male, 100
Event: 13-Feb-2018, Breast Stroke, Female, 100\
"""
data = eventsText.split("\n")
for i in data:
    thing = i.split(", ")
    thing[0] = thing[0][7:]
    statement = "INSERT INTO Events(date, `stroke type`,gender, distance) VALUES(?,?,?,?);"
    con.execute(statement,thing)

resultsText = """\
Results: 1, 1, 120.94
Results: 2, 1, 121.13
Results: 3, 1, 119.22
Results: 1, 3, 125.27
Results: 2, 3, 124.99
Results: 3, 3, 126.00
Results: 4, 2, 125.11
Results: 5, 2, 125.10
Results: 6, 2, 126.25
Results: 4, 4, 126.35
Results: 5, 4, 127.28
Results: 6, 4, 126.89\
"""

data = resultsText.split('\n')
for i in data:
    thing = i.split(", ")
    thing[0] = thing[0][9:]
    statement = "INSERT INTO Results(SwimmerID, EventID,time) VALUES(?,?,?);"
    con.execute(statement,thing)
```
