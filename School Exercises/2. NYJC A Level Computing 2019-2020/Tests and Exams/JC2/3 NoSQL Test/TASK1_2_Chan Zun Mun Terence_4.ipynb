```python
## Task 1.2

## Options
FILENAME = "STUDENTLIST.txt"
DATABASE_NAME = "all_classes"
COLLECTION_NAME = "student_details"

def list_coll(coll, query=None):
    output = ""
    cursor = coll.find(query)
    
    # Table Generation
    header = ["class", "index_no", "name", "remarks"]
    colwidth = [5,10,10,20]
    for j in range(len(header)):
        # Left align, matching col width
        output += f"{header[j]:<{colwidth[j]}} "
    output += "\n" # Newline
    
    # Show every element
    for i in cursor:
        for j in range(len(header)):
            # Left align
            value = str(i.get(header[j]))
            output += f"{value:<{colwidth[j]}} "
        output += "\n"
    print(output)
    
def list_class(coll, student_class):
    list_coll(coll, {"class": student_class})
    
def add_remarks(coll, student_class, student_index_no, 
                student_remarks):
    coll.update_one(
        {"class":student_class, "index_no":student_index_no}, # Query
        {"$set":{"remarks": student_remarks}} # Set the remarks
    )

def menu(coll):
    ## Main Loop
    continue_adding = True
    while continue_adding:
        
        ## Key in class
        student_class = input("Please enter a class: ")
        list_class(coll,student_class)
        
        ## Add Remarks
        student_index_no = int(input("Enter the index number of the student:"))
        student_remarks = input("Enter the remarks to be given/ updated: ")
        add_remarks(coll, student_class, student_index_no, student_remarks)
        print("Remarks Added/Updated\n")
        
        # Ask to continue?
        continue_input = input("Do you want to continue adding remarks for more students? Press N to exit: ")
        if continue_input == "N":
            continue_adding = False
    
    list_coll(coll)
            
import pymongo
conn = pymongo.MongoClient("127.0.0.1",27017)
db = conn.get_database(DATABASE_NAME)
coll = db.get_collection(COLLECTION_NAME)
menu(coll)
conn.close()
```
