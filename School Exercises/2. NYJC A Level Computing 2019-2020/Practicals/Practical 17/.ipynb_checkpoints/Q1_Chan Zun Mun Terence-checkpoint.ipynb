```
def pointerIncrement(pointer, capacity, increment=1):
    return (pointer + increment)  % (capacity)
class Queue:
    def __init__(self, capacity):
        self.items = [None] * capacity
        self.capacity = capacity
        self.count = 0
        self.front = 0
        self.rear = -1

    def isEmpty(self):
        return self.count == 0

    def isFull(self): 
        return self.capacity == self.count
    
    def enqueue(self, item):
        if self.isFull():
            print("The queue is full, cannot enqueue")
            return False
        else:
            self.rear = pointerIncrement(self.rear, self.capacity)
            self.items[self.rear] = item
            self.count += 1
            return True
        
    def dequeue(self):
        if self.isEmpty():
            print("The queue is empty, there is nothing to dequeue")
            return False
        else:
            front_item = self.items[self.front]
            self.items[self.front] = None
            self.front = pointerIncrement(self.front, self.capacity)
            self.count -= 1
            return front_item

    def display(self):
        # Constants
        start = 0
        end = self.capacity - 1
        # Variables
        output = ""
        pointer = start
        while pointer != end:
            if self.items == None:
                output += "None"
            else:
                output += str(i)
            output += ","
            pointer = pointerIncrement(pointer, self.capacity)
        if self.capacity > 0:
            output = output[:-1]
        return output
```

Write program code to define the classes Tree and Node.

a) Your program code. [16]


```
class Node:
    def __init__(self):
        self.key = 0
        # Initialise as None first to prevent excessive recursion
        self.left = None
        self.right = None
    
    def __str__(self):
        output = ""
        if self.left != None:
            output += str(self.left) + " "
        output += str(self.key)
        if self.right != None:
            output += " " + str(self.right)
        return output
        
    def insert(self, key:int):
        if key < self.key:
            if self.left == None:
                self.left = Node()
                self.left.key = key
            else:
                self.left.insert(key)
        elif key > self.key:
            if self.right == None:
                self.right = Node()
                self.right.key = key
            else:
                self.right.insert(key)

class Tree:
    def __init__(self):
        self.root = None # Node Type
    
    def add(self, newItem):
        if self.root == None:
            self.root = Node()
            self.root.key = newItem
        else:
            self.root.insert(newItem)


    def printTreeInOrder(self):
        '''
        Recursion implementation
        if self.root == None:
            print("")
        else:
            print(self.root)
        '''
        output = ""
        stack = []
        visited = set()
        stack.append(self.root)
        while len(stack) > 0:
            curr = stack[-1]
            visited.add(curr)
            # Simulate Left Tree Recursion
            if curr.left != None and curr.left not in visited: 
                stack.append(curr.left)
            else:
                stack.pop() #Done with node
                output += " " + str(curr.key)
                # Simulate Right Tree Recursion
                if curr.right != None: 
                    stack.append(curr.right)
        print(output[1:]) #Remove extra space at front
```

```
# Iterative implementation
def FindKthSmallest(k, tree):
    if tree.root == None:
        return None
    stack = []
    visited = set()
    stack.append(tree.root)
    while len(stack) > 0:
        curr = stack[-1]
        visited.add(curr)
        # Simulate Left Tree Recursion
        if curr.left != None and curr.left not in visited: 
            stack.append(curr.left)
        else:
            stack.pop() #Done with node
            k -= 1
            if k == 0:
                return curr.key
            # Simulate Right Tree Recursion
            if curr.right != None: 
                stack.append(curr.right)
    return None
```

```
# Morris implementation
def FindKthSmallest(k, tree):
    if tree.root == None:
        return None
    curr = tree.root
    prev = None
    while curr != None:
        # Left Dead End
        if curr.left == None:
            # Could be small element
            k -= 1
            value = curr.key
            if k == 0: 
                return value
            
            prev = curr
            curr = curr.right
        else:
            # Find inorder predecessor, Traverse right
            prev = curr.left
            while (prev.right != None and prev.right != curr):
                prev = prev.right
            
            # Set head node
            if prev.right == None:
                prev.right = curr
                curr = curr.left
            else: # Remove Link
                prev.right = None
                # Could be small element
                k -= 1
                if k == 0: 
                    return value
                value = curr.key
                curr = curr.right
    return None
```
