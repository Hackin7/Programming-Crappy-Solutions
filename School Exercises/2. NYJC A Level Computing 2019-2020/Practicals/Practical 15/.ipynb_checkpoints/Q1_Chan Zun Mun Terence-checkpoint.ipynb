```
class Node:
  def__init__(self, data):
      self.data = data
      self.link = None

class LinkedList:
  def __init__(self):
    self.head = None
  def insert(self, data):
    # create new node
    new_node = Node(data)

    # if linkedlist empty
    if self.head == None:
      self.head = new_node
    else:
      curr = self.head
      prev = None
      while curr != None and data > curr.data: # ascending sort
        prev = curr
        curr = curr.link
      
      if prev == None: # insert to front
        new_node.link = curr 
        self.head = new_node
      elif curr == None: # insert to end
        new_node.link = None 
        prev.link = new_node
      else: # insert in between
        new_node.link = curr
        prev.link = new_node
```
