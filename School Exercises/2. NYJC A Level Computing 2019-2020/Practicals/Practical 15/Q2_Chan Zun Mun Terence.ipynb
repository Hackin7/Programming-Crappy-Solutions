```python
class Node:
  def __init__(self, data):
      self.data = data
      self.link = None

class LinkedList:
  def __init__(self):
    self.head = None

  def insert(self, data):
    # create new node
    new_node = Node(data)

    # if linkedlist empty
    if self.head == None:
      self.head = new_node
    else:
      curr = self.head
      prev = None
      while curr!=None and data > curr.data: # Ascending sort
        prev = curr
        curr = curr.link
      
      if prev == None: # insert to front
        new_node.link = curr 
        self.head = new_node
      elif curr == None: # insert to end
        new_node.link = None 
        prev.link = new_node
      else: # insert in between
        new_node.link = curr
        prev.link = new_node

  def delete(self, data):
    # Traverse through Linked List
    curr = self.head
    prev = None
    while curr != None and data > curr.data:
      prev = curr
      curr = curr.link
    
    if curr != None and curr.data == data:
      if prev == None: # At the start
        self.head = curr.link
      else:
        prev.link = curr.link #Skip old node

  def update(self, old_data, new_data):
    if self.search(old_data):
        # Delete old data
        self.delete(old_data)
        # Insert new data
        self.insert(new_data)

  def search(self, data) -> bool:
    # Traverse through Linked List
    curr = self.head
    prev = None
    while curr != None and data > curr.data:
      prev = curr
      curr = curr.link
    
    if curr != None and curr.data == data:
      return True
    else:
      return False

  def display(self):
    output = ""
    # Traverse through Linked List
    curr = self.head
    prev = None
    while curr != None:
      output += str(curr.data)+", "
      prev = curr
      curr = curr.link
    #Display to screen
    if output != "":
      output = output[:-2]
    print(output)
```

```python
# Driver Code
lis = LinkedList()

lis.insert(2) # Insert at front
lis.insert(3) # Insert at end
lis.insert(0) # Insert at front
lis.insert(1) #Insert in between
lis.display()

lis.delete(-1) #Remove non existant member at front
lis.delete(5) #Remove non existant member at end
lis.delete(1) # Remove element in between
lis.delete(0) # Remove element at front
lis.delete(3) # Remove element at end
lis.display()

# Reset to original data
lis.insert(3)
lis.insert(0)
lis.insert(1) 
lis.display()
# Update
lis.update(-1, 0) #Non existant member at front
lis.update(5, 0) #Non existant member at end
lis.update(1, 1.1) # Element in between
lis.update(0, 2) # Element at front
lis.update(3, 4.3) # Element at end
lis.display()

# Reset to original data
lis.insert(3)
lis.insert(0)
lis.insert(1) 
#Search
print(lis.search(-1)) #Non existant member at front
print(lis.search(5)) #Non existant member at end
print(lis.search(1)) # Element in between
print(lis.search(0)) # Element at front
print(lis.search(3)) # Element at end
```
