```python
# Task 4.1

def tableInput():
    dataDictionary = {}
    
    keepRunning = True
    while keepRunning:
        X = input('Next X value ... <ZZZ to END> ')
        if X == "ZZZ":
            keepRunning = False
        else:
            frequency = int(input("Frequency ... "))
            dataDictionary[X] = frequency
    return dataDictionary

def displayData(dataDictionary):
    output = ""
    
    # Heading
    output += "+" * 40 
    output += "\n"
    output += "Frequency distribution"
    output += "\n"
    output += "+" * 40
    output += "\n"
    
    # Table
    for X, frequency in dataDictionary.items():
        output += "\n"
        output += f"  {X:<8}"
        output += "@" * frequency # Bar
        
    print(output)
    
### Main ####################################################
'''
Data
XS, 12
S, 22
M, 45
L, 56
XL, 39
XXL, 11
'''
def task4_1():
    dataDictionary = tableInput()
    print()
    print()
    displayData(dataDictionary)
    
task4_1()
```

```python
# Task 4.3

def bar(freq, scaling=1, barChar="#"):
    lengthOfBar = (freq // scaling) + (freq % scaling > 0)
    return barChar * lengthOfBar
    
def suitableScale(generateOutput, maxWidth=80, **kwargs):
    scale = 1
    fit = False
    while not fit:
        ### Generate width ###########################
        lines = generateOutput(scale=scale, **kwargs).split('\n')
        width = 0
        for line in lines:
            width = max(len(line), width)
        ######################################
        if width > maxWidth:
            scale += 1
        else:
            fit = True
    return scale
    
def generateDataOutput(dataDictionary, scale=1, barWidth=1, barChar='#'):
    output = ""
    
    '''
    ### Set bar width such that lines fit ##########################
    fit = False
    while not fit:
        noLines = 4 + len(dataDictionary.items()) * barWidth + 3
        if noLines > 40:
            barWidth -= 1
        else:
            fit = True
    
    ### Set scale such that lines fit ##########################
    maxFreq = max(dataDictionary.values())
    scale = 1
    fit = False
    while not fit:
        width = 10 + (maxFreq // scale) + (maxFreq % scale) + 10
        if width > 80:
            scale += 1
        else:
            fit = True
    '''
    
    ### Output #####################################################
    # Heading
    output += "+" * 40 
    output += "\n"
    output += "Frequency distribution"
    output += "\n"
    output += "+" * 40
    output += "\n"
    
    # Table
    for X, frequency in dataDictionary.items():
        output += "\n"
        output += f"  {X:<8}" # 10 characters
        output += bar(frequency, scale, barChar) # Bar
        output += ' ' * 10
        for i in range(barWidth - 1):
            output += "\n"
            output += f"  {' '*8:<8}" # Padding
            output += bar(frequency, scale, barChar) # Bar
            output += ' ' * 10
        
    # Scaling
    output += "\n"
    output += " " * 10 + '----|' * 11 + '-----' 
    output += "\n"
    # Labelling
    output += " " * 10
    for i in range(11):
        position = scale * 5 * (i + 1)
        output += f"{position:5}"
    
    return output
    
### Main ###############################################################
dataset1 = {
    'XS': 12,
    'S': 22,
    'M': 45,
    'L': 56,
    'XL': 39,
    'XXL': 11
}

dataset2 = { # Have to scale down by 2
    'XS': 24,
    'S': 44,
    'M': 90,
    'L': 112,
    'XL': 78,
    'XXL': 22
}
dataset3 = { # Have to scale down by 3, with remainders on XXL
    'XS': 36,
    'S': 66,
    'M': 135,
    'L': 168,
    'XL': 117,
    'XXL': 34
}

def datasetTester(name, dataset):
    print("Dataset", name)
    barWidth = suitableWidth(generateDataOutput, 2, dataDictionary=dataset)
    scale = suitableScale(generateDataOutput, dataDictionary=dataset)
    output = generateDataOutput(dataset, scale, barWidth, '0')
    output = lineWidthPadding(output) 
    #print(len(output.split("\n")), min([len(x) for x in output.split('\n')])) # Debugging: No lines, line width
    print(output)
    print()
    
def task4_3():
    datasetTester(1, dataset1)
    datasetTester(2, dataset2)
    datasetTester(3, dataset3)
    
task4_3()
```
