```python
## Task 3.1, Evidence 6
class Node:
    def __init__(self):
        self.data = None
        self.leftPtr = None
        self.rightPtr = None
    
    def constructor(self):
        self.__init__(self)
    
    # Setter functions
    def setData(self, s:str):
        self.data = s
    
    def setLeftPtr(self, x:int):
        self.leftPtr = x
        
    def setRightPtr(self, y:int):
        self.rightPtr = y
    
    # Getter functions
    def getData(self)->str:
        return self.data
    
    def getLeftPtr(self)->int:
        return self.leftPtr
    
    def getRightPtr(self)->int:
        return self.rightPtr
    
class Tree:
    def __init__(self):
        # Generate an array
        self.tree = [] # Python list simulate an array
        arraySize = 7 # Fixed array size 
        for i in range(arraySize): # Insert placeholder items
            self.tree.append(None)
            
        self.root = -1
    
    def constructor(self):
        self.__init__(self)
    
    def add(self, newItem): # newItem is assumed to be a string
        pos = 0 # New item position
        
        if self.root == -1: # No items in tree
            self.root = 0
        else: # Traverse the tree
            found = 0
            for node in range(len(self.tree)):
                if self.tree[node] == None and not found:
                    pos = node
                    found = True
                    
            # Link new item
            current = self.root
            while current != None:
                # Use python's built-in string comparison
                if newItem < self.tree[current].getData(): 
                    ptr = self.tree[current].getLeftPtr()
                    if ptr == None:
                        self.tree[current].setLeftPtr(pos)
                        current = None # Exit
                    else:
                        current = ptr
                else: 
                    ptr = self.tree[current].getRightPtr()
                    if ptr == None:
                        self.tree[current].setRightPtr(pos)
                        current = None # Exit
                    else:
                        current = ptr
        # Create new item
        self.tree[pos] = Node()
        self.tree[pos].setData(newItem)
        
    def print(self): # Iterative implementation of BFS
        output = ""
        
        stack = []
        stack.append(self.root)
        # Directed graph, so no need for visited array
        while len(stack) > 0:
            # Getting top element
            current = self.tree[stack.pop(0)]
            # Printing
            output += current.getData() + "\n"
            if current.getLeftPtr() != None:
                stack.append(current.getLeftPtr())
            if current.getRightPtr():
                stack.append(current.getRightPtr())
        
        print(output)
```
