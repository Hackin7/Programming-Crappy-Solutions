```python
## Task 3.4
## Evidence 9 and 10
def readFileToLinkedList(filename, linkedList):
    file = open(filename,'r') # Read mode
    for line in file:
        line = line.strip() #Remove newline
        data = line.split(",")
        
        NewDataItem = data[0]
        ParentItem = data[1]
        ThisMove = data[2]
        
        linkedList.AddToRobotData(NewDataItem, ParentItem, ThisMove)
    file.close()

linkedList = LinkedList()
readFileToLinkedList("SEARCHTREE.txt", linkedList)
linkedList.OutputData()
```
