```python
## Task 4.1
## Evidence 13

def initialisePuzzle():
    # Nested lists to represent the puzzle
    # Puzzle initalised with values shown
    puzzle = [
        [4, 3, 2, 1],
        [1, 2, 4, 3],
        [3, 4, 1, 2],
        [2, 1, 3, 4]
    ]
    return puzzle
```

```python
## Task 4.2
## Evidence 14 and Evidence 15

def displayPuzzle(puzzle):
    output = ""
    
    for row in puzzle:
        for item in row:
            output += str(item) + " "
        # Remove trailing whitespace and add new line
        output = output[:-1]+"\n"
    
    print(output)
    
puzzle = initialisePuzzle()
displayPuzzle(puzzle)
```

```python
## Task 4.3
## Evidence 16 and 17

## Helper functions
def swapRows(puzzle, row1, row2):
    temp = puzzle[row1]
    puzzle[row1] = puzzle[row2]
    puzzle[row2] = temp

def swapColumns(puzzle, col1, col2):
    # Swap the 2 column values for each row
    for row_index in range(len(puzzle)):
        temp = puzzle[row_index][col1]
        puzzle[row_index][col2] = puzzle[row_index][col1]
        puzzle[row_index][col2] = temp

## Transformations

# Swaps two rows in the same quadrants
def transformation1(puzzle, switchUpper=True):
    if switchUpper:
        row1 = 0
        row2 = 1
    else:
        row1 = 2
        row2 = 3
    swapRows(puzzle, row1, row2)
    
    
# Swaps two columns in the same quadrants
def transformation2(puzzle, switchLeft=True):
    if switchLeft:
        col1 = 0
        col2 = 1
    else:
        col1 = 2
        col2 = 3
    swapColumns(puzzle, col1, col2)
    
    
# Swaps the left and right quadrant rows entirely
def transformation3(puzzle):
    swapRows(puzzle, 0, 2)
    swapRows(puzzle, 1, 3)
    
def transformation4(puzzle):
    swapColumns(puzzle, 0, 2)
    swapColumns(puzzle, 1, 3)

# # List of transformations, 
#  in the format of (transform_function, needRandomise:bool, commentary)
transformations = [
    (transformation1, 1, "Transformation 1: Swaps two rows in the same quadrants"),
    (transformation2, 1, "Transformation 2: Swaps two columns in the same quadrants"),
    (transformation3, 0, "Tranformation 3: Swaps the left and right quadrant rows entirely"),
    (transformation4, 0, "Tranformation 4: Swaps the left and right quadrant columns entirely")
]

from random import randint
def task4_3():
    ## Initalise puzzle
    puzzle = initialisePuzzle()
    displayPuzzle(puzzle)
    
    ## Randomly choose transformations
    choices = []
    for choice in range(2):
        index = randint(0,len(transformations)-1)
        choices.append(transformations[index])
        transformations.pop(index)
    
    ## Show Transformations
    for transform_function, needRandomise, commentary in choices:
        if needRandomise:
            transform_function(puzzle, randint(0,1))
        else:
            transform_function(puzzle)
        
        print(commentary)
        displayPuzzle(puzzle)

task4_3()
```
