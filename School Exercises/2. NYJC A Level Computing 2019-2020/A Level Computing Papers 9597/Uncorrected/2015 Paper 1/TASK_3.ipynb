```python
## Task 3.1, Evidence 9

import random
def LicenceKey() -> str: # Generates a new licence key
    # Generate 9 random characters
    key = ""
    total = 0
    for multiplier in range(1, 9+1):
        asciiCode= 64 + random.randint(1, 26)
        product = asciiCode * multiplier
        total += product
        key += chr(asciiCode)
        
    # Generate checkdigit
    checkDigit = total % 11
    if checkDigit == 10:
        checkCharacter = "X"
    else:
        checkCharacter = str(checkDigit)
    key += checkCharacter
    
    return key

for i in range(3):
    print(f"Random Licence Key {i+1}: {LicenceKey()}")
```

```python
## Task 3.4, Evidence 11
'''
Test Cases
1. Input: ADMTZFUOM5 
  - Test to check if it works on normal data
  - Expected output: ERROR: This licence key is only a single user licence, cannot add registered users
2. Input: TYIOCQKYC4
  - Test to check if invalid key shows appropriate message
  - Expected output: ERROR: This licence key is only a single user licence, cannot add registered users
3. Input: 1
  - Test on abnormal data, to see if it can detect a key doesn't exist
  - Expected Output: ERROR: No such Licence Key exists
'''
def option2():
    key = input("Enter the Licence Key to add to:")
    # Find and edit entry
    newContents = ""
    
    added = False
    file = open(FILENAME)
    for line in file:
        line = line.strip() #Remove trailing newlines
        data = line.split(" ")
        if data[0] == key:
            if len(data) == 3 and data[1] == "3":
                if int(data[2]) >= 3:
                    print("ERROR: This licence key had reached its maximum capacity of 3 users")
                    return
                else:
                    data[2] = str(int(data[2])+1)
                    added = True
                    newContents = newContents.strip()
                    newContents += f"\n{data[0]} {data[1]} {data[2]}\n"
            else:
                print("ERROR: This licence key is only a single user licence, cannot add registered users")
                return
        if not added:
            newContents += line + "\n"
    
    if not added:
        print("ERROR: No such Licence Key exists")
        return
    else: 
        print("Added!")
    file.close()
    
    # Overwrite data
    file = open(FILENAME, "w")
    file.write(newContents)
    file.close()
    
menu()
```

```python
# Task 3.5, Evidence 12
from datetime import datetime
class Licence:
    def __init__(self):
        self.licenceKey = ""
        self.purchaseDate = None
        self.purchaser = "" 
        self.type = ""

    ### Getter ##############################
    def getLicenceKey(self):
        return self.licenceKey
        
    def getPurchaseDate(self) -> datetime:
        return self.purchaseDate
        
    def getPurchaser(self):
        return self.purchaser
    
    def getType(self):
        return self.type
        
    ### Setter ###############################
    def setLicenceKey(self, val):
        self.licenceKey = val
    
    def setPurchaseDate(self, val: datetime):
        self.purchaseDate = val
        
    def setPurchaser(self, val):
        self.purchaser = val
    
    def setType(self, val):
        self.type = val
        
class SingleUserRegistration(Licence):
    def __init__(self):
        Licence.__init__(self)
        self.type = "Single-user Registration"
        self.registrationDate = None
        self.registeredMACAddress = ""
    
    ### Getter ##############################
    def getRegistrationDate(self) -> datetime:
        return self.registrationDate
        
    def getRegisteredMACAddress(self):
        return self.purchaser
        
    ### Setter ###############################
    def setRegistrationDate(self, val: datetime):
        self.registeredMACAddress = val
        
    def setRegisteredMACAddress(self, val):
        self.registeredMACAddress = val
        
class ThreeUserRegistration(Licence):
    def __init__(self):
        Licence.__init__(self)
        self.type = "3-user Registration"
        self.registrationDate = [None, None, None]
        self.registeredMACAddress = [None, None, None]
        self.registered = 0
    ### Getter ##############################
    def getRegistrationDate(self, index) -> datetime:
        return self.registrationDate[index]
        
    def getRegisteredMACAddress(self, index):
        return self.registeredMACAddress[index]
        
    ### Setter ###############################
    def setRegistrationDate(self, index, val: datetime):
        self.purchaseDate = val
        
    def setRegisteredMACAddress(self, index, val):
        self.registeredMACAddress = val
    
    def addRegistered(date, macAddress):
        index = self.registered
        self.setRegistrationDate(index, date)
        self.setRegisteredMACAddress(index, macAddress)
        self.registered += 1
```
