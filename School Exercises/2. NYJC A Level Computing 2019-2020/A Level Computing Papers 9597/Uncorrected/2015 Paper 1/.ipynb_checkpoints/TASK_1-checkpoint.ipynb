```python
#Task 1.1, Evidence 1

MENU_TEXT = """\
1. Read file data
2. Bubble sort
3. Quick sort
4. End"""

def menu():
    Admissions = []
    
    option = 0 # Invalid option makes the program loop again
    while option != 4:
        print(MENU_TEXT)
        option = int(input("Enter option: "))
        if option == 1:
            Admissions = readFileData()
        elif option == 2:
            Admissions = bubbleSort(Admissions)
            display(Admissions)
        elif option == 3:
            Admissions = quickSort(Admissions)
            display(Admissions)
```

```python
## Task 1.2
            
def readFileData(filename = "ADMISSIONS-DATA.TXT"):
    Admissions = []
    file = open(filename, "r") # Read mode
    for line in file:
        Admissions.append(int(line))
    file.close()
    return Admissions
    
def display(Admissions):
    print(f"Admissions data: {Admissions}")

## Evidence 2
def bubbleSort(Admissions):
    UpperBound = len(Admissions)
    NoSwaps = False
    while NoSwaps == False:
        NoSwaps = True
        for Posn in range(0, UpperBound - 1):
            if Admissions[Posn] > Admissions[Posn + 1]:
                NoSwaps = False
                Temp = Admissions[Posn]
                Admissions[Posn] = Admissions[Posn + 1]
                Admissions[Posn + 1] = Temp
    return Admissions

## Evidence 3
# An insertion sort is used
def quickSort(Admissions):
    ## Base case
    if len(Admissions) <= 1:
        return Admissions
    
    pivot = Admissions[0]
    left = []
    right = []
    for index in range(1, len(Admissions)):
        if Admissions[index] < pivot:
            left.append(Admissions[index])
        else:
            right.append(Admissions[index])
        
    left.append(pivot)
    left = quickSort(left)
    right = quickSort(right)
    Admissions = left + right
    #print(left, right, Admissions)
    return Admissions

menu()
```

```python
lis = [0,4,0,4,4]
lis = quickSort(lis)
print(lis)
```

```python
## Task 1.3, Evidence 4
def bubbleSort(Admissions):
    UpperBound = len(Admissions)
    NoSwaps = False
    comparisons = 0
    while NoSwaps == False:
        NoSwaps = True
        for Posn in range(0, UpperBound - 1):
            comparisons += 1
            if Admissions[Posn] > Admissions[Posn + 1]:
                NoSwaps = False
                Temp = Admissions[Posn]
                Admissions[Posn] = Admissions[Posn + 1]
                Admissions[Posn + 1] = Temp
    print(f"Number of comparisons for bubble sort: {comparisons}")
    return Admissions
    
# An insertion sort is used
def quickSort(Admissions, showComparisons=True):
    comparisons = 0
    ## Base case
    if len(Admissions) <= 1:
        return Admissions, comparisons
    
    pivot = Admissions[0]
    left = []
    right = []
    for index in range(1, len(Admissions)):
        comparisons += 1
        if Admissions[index] < pivot:
            left.append(Admissions[index])
        else:
            right.append(Admissions[index])
        
    left.append(pivot)
    
    ## Recursion
    side = quickSort(left, False)
    left = side[0]
    leftComparisons = side[1]
    
    side = quickSort(right, False)
    right = side[0]
    rightComparisons = side[1]
    
    Admissions = left + right
    comparisons += leftComparisons + rightComparisons
    
    if showComparisons:
        print(f"Number of comparisons for quick sort: {comparisons}")
        return Admissions
    else:
        return (Admissions, comparisons)

menu()
```
