```python
## Task 1.1 
## Evidence 1 and Evidence 2

def readContents(filename):
    Inventory = [] # List used
    
    file = open(filename,"r") # Open file in read mode
    for item in file:
        Inventory.append(item.strip())
    file.close()
    
    return Inventory

def getItemTypes(Inventory):
    ItemTypes = [] # List used
    
    for item in Inventory:
        # Generic Searching algorithm
        inItemTypes = False
        for itemType in ItemTypes:
            if item == itemType:
                inItemTypes == True
        
        # Add if there are no duplicates
        if not inItemTypes:
            ItemTypes.append(item)
            
    return ItemTypes

def getItemCounts(Inventory, ItemTypes):
    ItemCounts = {} # Python Dictionary used
    
    # Generate ItemCounts
    for itemType in ItemTypes:
        ItemCounts[itemType] = 0
    
    for item in Inventory:
        ItemCounts[item] += 1 # Add evert item
        
    return ItemCounts

def showItemCounts(ItemCounts):
    output = ""
    
    output += f"{'ItemType':<18}{'Count'}\n\n"
    
    for item in ItemCounts:
        count = ItemCounts[item]
        output += f"{item:<18}{count}\n"
    return output


## Main Program Code
Inventory = readContents("INVENTORY.txt")
ItemTypes = getItemTypes(Inventory)
ItemCounts = getItemCounts(Inventory, ItemTypes)
output = showItemCounts(ItemCounts)

print(output)
```
