```python
## Task 3.4
## Evidence 9 and 10
def readFileToLinkedList(filename, linkedList):
    file = open(filename,'r') # Read mode
    for line in file:
        line = line.strip() #Remove newline
        data = line.split(",")
        
        NewDataItem = data[0]
        ParentItem = data[1]
        ThisMove = data[2]
        
        linkedList.AddToRobotData(NewDataItem, ParentItem, ThisMove)
    file.close()

linkedList = LinkedList()
readFileToLinkedList("SEARCHTREE.txt", linkedList)
linkedList.OutputData()
```

```python
## Task 3.5
## Evidence 11 and 12

class LinkedList:
    def __init__(self):
        self.RobotData = []
        # 1 indexed, the 0th index is excluded
        for index in range(25+1):
            self.RobotData.append(ConnectionNode(None,index+1,0))
        self.Root = 1 # Index for the root position of the RobotData array
        self.NextFreeChild = 1 #Index for next available empty node
    
    ### Evidence 7 ##########################
    def FindNode(self,NodeValue): # Returns int
        Found = False
        CurrentPosition = self.Root
        while not(Found == True or CurrentPosition > 25):
            if self.RobotData[CurrentPosition].DataValue == NodeValue :
                Found = True
            else:
                CurrentPosition = CurrentPosition + 1
        if CurrentPosition > 25 :
            return 0
        else:
            return CurrentPosition

    def AddToRobotData(self,NewDataItem, ParentItem, ThisMove):
        if self.Root == 1 and self.NextFreeChild == 1 :
            self.NextFreeChild = self.RobotData[self.NextFreeChild].LeftChild
            self.RobotData[self.Root].LeftChild = 0
            self.RobotData[self.Root].DataValue = NewDataItem
        else:
            # does the parent exist?
            ParentPosition = self.FindNode(ParentItem)
            if ParentPosition > 0 :  # parent exists
                # does the child exist?
                ExistingChild = self.FindNode(NewDataItem)
                if ExistingChild > 0 : # child exists
                    ChildPointer = ExistingChild
                else:
                    ChildPointer = self.NextFreeChild
                    self.NextFreeChild = self.RobotData[self.NextFreeChild].LeftChild
                    self.RobotData[ChildPointer].LeftChild = 0
                    self.RobotData[ChildPointer].DataValue = NewDataItem
                if ThisMove == 'L' :
                    self.RobotData[ParentPosition].LeftChild = ChildPointer
                else:
                    self.RobotData[ParentPosition].RightChild = ChildPointer
    
    ### Evidence 8 ###############################
    def OutputData(self):
        output = ""
        for index in range(1,len(self.RobotData)):
            output += str(self.RobotData[index]) + "\n"
        print(output)
    
    ### Evidence 11 ############################### OVERHAUL CORRECTIONS
    def PreOrderTraversalRoutes(self, NodeIndex=None, currentRoute = [], end='Z'):
        if NodeIndex == None:
            NodeIndex = self.Root # Start from A
            
        currentRoute = currentRoute + [self.RobotData[NodeIndex].DataValue] # Copy the list
        ##
        if self.RobotData[NodeIndex].DataValue == end:
            output = ""
            for node in currentRoute:
                output += node + "->"
            print(output[:-2])
            return

        ## If there is a left child
        leftNodeIndex = self.RobotData[NodeIndex].LeftChild
        if leftNodeIndex != 0:
            self.PreOrderTraversalRoutes(self.RobotData[NodeIndex].LeftChild, currentRoute)
        
        ## If there is a right child
        rightNodeIndex = self.RobotData[NodeIndex].RightChild
        if rightNodeIndex != 0:
            self.PreOrderTraversalRoutes(self.RobotData[NodeIndex].RightChild, currentRoute)
        
linkedList = LinkedList()
readFileToLinkedList("SEARCHTREE.txt", linkedList)
linkedList.PreOrderTraversalRoutes()
```
