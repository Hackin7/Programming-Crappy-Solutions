```python
## Task 3.1, Evidence 6
class Node:
    def __init__(self):
        self.__data = "" # Represent none value
        self.__leftPtr = -1
        self.__rightPtr = -1
    
    def constructor(self):
        self.__init__(self)
    
    # Setter functions
    def setData(self, s:str):
        self.__data = str(s)
    
    def setLeftPtr(self, x:int):
        self.__leftPtr = int(x)
        
    def setRightPtr(self, y:int):
        self.__rightPtr = int(y)
    
    # Getter functions
    def getData(self)->str:
        return self.__data
    
    def getLeftPtr(self)->int:
        return self.__leftPtr
    
    def getRightPtr(self)->int:
        return self.__rightPtr
    
class Tree:
    def __init__(self):
        # Generate an array
        self.__tree = [] # Python list simulate an array
        arraySize = 7 # Fixed array size 
        for i in range(arraySize): # Insert placeholder items
            self.__tree.append(Node())
            
        self.__root = -1
    
    def constructor(self):
        self.__init__(self)
    
    def add(self, newItem): # newItem is assumed to be a string
        pos = 0 # New item position
        
        if self.__root == -1: # No items in tree
            self.__root = 0
        else: # Traverse the tree
            found = 0
            for node in range(len(self.__tree)):
                if self.__tree[node].getData() == "" and not found:
                    pos = node
                    found = True
                    
            # Link new item
            current = self.__root
            while current != -1:
                # Use python's built-in string comparison
                if newItem < self.__tree[current].getData(): 
                    ptr = self.__tree[current].getLeftPtr()
                    if ptr == -1:
                        self.__tree[current].setLeftPtr(pos)
                        current = -1 # Exit
                    else:
                        current = ptr
                else: 
                    ptr = self.__tree[current].getRightPtr()
                    if ptr == -1:
                        self.__tree[current].setRightPtr(pos)
                        current = -1 # Exit
                    else:
                        current = ptr
        # Create new item
        #self.__tree[pos] = Node()
        self.__tree[pos].setData(newItem)
        
    def print(self):
        output = ""
        output += f"{'Data':5}{'Left':5}{'Right':5}\n"
        for current in self.__tree:
            output += f"{current.getData():5}{current.getLeftPtr():5}{current.getRightPtr():5}\n"
        print(output)    

            
    def printBFS(self): # Iterative implementation of BFS
        output = ""
        stack = []
        stack.append(self.__root)
        # Directed graph, so no need for visited array
        while len(stack) > 0:
            # Getting top element
            current = self.__tree[stack.pop(0)]
            # Printing
            output += current.getData() + "\n"
            if current.getLeftPtr() != -1:
                stack.append(current.getLeftPtr())
            if current.getRightPtr() != -1:
                stack.append(current.getRightPtr())
        
        print(output)
```
